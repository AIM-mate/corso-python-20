##Ciclo While

```
i = 1

while i < 5:
    print(i)
    
    i += 1
    
print("done")
```

Quando si scrive un ciclo while è fondamentale ricordarsi di aggiornare l'iteratore, in modo da poter uscire dal ciclo.

Si può scegliere di impostare una condizione sempre vera, a patto di aggiungere una sequenza di uscita dal ciclo

```
while True:
  parola = input("Inserisci qualcosa: ")
  if parola == "stop":
    break
  print(parola)
print("Finito")
```

A volta può essere necessario saltare alcune delle istruzioni contenute all'interno del ciclo e passare all'iterazione successiva. Per questo usiamo il comando continue

```
while True:
  parola = input("Inserisci qualcosa: ")
  if parola == "spam":
    continue
  if parola == "stop":
    break
  print(parola)
print("Finito")
```

## Massimo e minimo

Possiamo usare un ciclo for e una variabile di tipo None per cercare il massimo e minimo valore di una lista.


```
massimo = None
minimo = None
print("All'inizio: ", "Massimo =", massimo, "Minimo =", minimo)
for numero in [34, 67, 1, 24, -5, 89, 16]:
  if massimo is None or numero > massimo:
    massimo = numero
  if minimo is None or numero < minimo:
    minimo = numero
  print("Attuale =",numero, "Massimo =", massimo, "Minimo =", minimo)
print("Massimo =", massimo, "Minimo =", minimo)
```

# Tuple

Una tupla è una sequenza di valori che, a differenza di una lista, è immutabile

```
# definizione, prima differenza con le liste: le parentesi tonde!!!
tupla = ("Jean Paul", "Gianmarco", "Luca")
#Le parentesi non sono necessarie, ma molto utili per la leggibilità
print(tupla)
```

```
print(tupla[1])
```

```
# seconda differenza con le liste: le tuple non si toccano!!!
tupla = ("Jean Paul", "Gianmarco", "Luca")
tupla[0]="Gian Paolo"
```

```
# come aggirare il problema?
tupla = ("Jean Paul", "Gianmarco", "Luca")
lista=list(tupla)
lista[0]="Gian Paolo"
tupla1=tuple(lista)
print(tupla1)
```

Molti degli operatori visti finora funzionano anche con le tuple, come le parentesi quadre per estrarre un elemento e gli operatori di slicing

Sulle tuple si possono utilizzare gli operatori di confronto: Python confronterà il primo elemento con il primo. Se sono uguali procederà ai secondi, altrimenti restituirà il risultato del confronto tra i due,

```
(0,1,1,4000000000) < (0,1,2,0.05)
```

Le tuple permettono di assegnare valori a più variabili contemporaneamente


```
m = ["uno", "due"] #Lista di due elementi
x , y = m
print(x)
print(y)

#Un'altra sintassi valida sarebbe con (x,y) al posto di x,y
addr = 'monty@python.org'
(nome , dominio) = addr.split('@')
print(nome, dominio)
```

```
l1 = [1, 2, 3]
l2 = ['a', 'b', 'c']
ll=zip(l1, l2)
print(ll)

for i, (x, y) in enumerate(ll):
    print(i,x,y)
```

#Dizionari

Un dizionario, come una lista, lega un insieme di indici (keys) a un insieme di valori. Al contrario delle liste, però, gli indici non sono necessariamente dei numeri, e non vi è quindi un ordine tra gli elementi.

```
#definizone
dizionario_AIM = dict()
dizionario_AIM = {
  "Presidente": "JPGB",
  "Vice": "Luca",
  "Team informatico": "Gianmarco",
  "Team Eventi": "Angelica",
  "Numero di soci": 20000
}
print(dizionario_AIM )
```

```
chiavi = ["Anno", "Mese", "Numero", "Giorno"]
valori = [1998, "Settembre", 5, "Sabato"]

d = { chiave:valore for (chiave,valore) in zip(chiavi, valori)}
d
```

```
#Accesso a un dizionario
print(dizionario_AIM["Presidente"])
print(dizionario_AIM["Numero di soci"])
```

```
#modifica di un dizionario
dizionario_AIM["Numero di soci"]=1718
print(dizionario_AIM["Numero di soci"])
```

```
#rimuovi elemento
dizionario_AIM.pop("Team informatico")

del dizionario_AIM["Presidente"]
print(dizionario_AIM)
```

```
#aggiungi elemento
dizionario_AIM['Team Accademico'] = 'GIULIA_MESC'
print(dizionario_AIM)
```

Possiamo usare la funzione len anche sui dizionari. Restituisce il numero di keywords disponibili

```
len(dizionario_AIM)
```

Anche per i dizionari, possiamo utilizzare l'operatore in per sapere se detrminati valori appartengono o meno al dizionario

```
valori = list(dizionario_AIM.values())
"JPGB" in valori
chiavi = list(dizionario_AIM.keys())
"Team amministrativo" in chiavi

```

Il metodo get: riceve in input una chiave e un valore di defaul. Se la chiave è presente restituisce il valore assegnatole, altrimenti quello di default

```
#Contare le lettere di una parola con un dizionario
parola = "Hipopotomonstrosesquipedaliofobia"
d = dict ()
for c in parola.lower():
  d[c] = d.get(c,0) + 1
print(d)
```

Cicli nei dizionari

```
#Possiamo ciclare attraverso il dizionario in maniera casuale
for chiave in d:
  print(chiave, d[chiave])

#Oppure possiamo convertire le chiavi in una lista e ordinarla
lst = list(d.keys())
lst.sort()
for chiave in lst:
  print(chiave, d[chiave])
```

Nota: l'operatore in usa degli algoritmi diversi per le liste e per i dizionari. Questo fa sì che il tempo impiegato per cercare un elemento in una lista cresce linearmente con la lunghezza, mentre si mantiene all'incirca costante in un dizionario

#File, introduzione basica

Apriamo e scriviamo su un file

```
new_file = open('new', 'w')
new_file.write('ciao')
new_file.close()

romeo = open('romeo.txt','r')
linea = romeo.readline()
linee = romeo.readlines()
print(linea)
```

```
name = input('Enter file:')
handle = open(name, 'r')
counts = dict()

for line in handle:
    words = line.split()
    for word in words:
        counts[word] = counts.get(word, 0) + 1

bigcount = None
bigword = None
for word, count in list(counts.items()):
    if bigcount is None or count > bigcount:
        bigword = word
        bigcount = count

print(bigword, bigcount, counts)
```

#Esercizietti

Esercizietto: Dato un numero, invertilo senza usare stringhe e indica se è palindromo

```
number = 12345

inverso = 0
while number > 0:
  cifra = number % 10
  inverso = inverso * 10 + cifra
  number = number // 10

print(inverso)
print(inverso == number)
```

Esercizietto: Dato un numero, indica se è primo

```
numero = 1347
primo = True

for i in range (2, numero // 2 +1):
  if number % i == 0:
    primo = False
    break

print(primo)
```
