```python
# PROBLEMA
# ricevo in input, **uno alla volta**, una sequenza di numeri, voglio creare una sequenza di numeri dove prima ci siano tutti quelli negativi, 
# e dopo tutti gli zeri e quelli positivi, non mi interessa del loro ordine. 
# Se ricevessi direttamente una lista di numeri potrei semplicemente ordinarla, è importante che i numeri siano intesi come
# stream continuo e letti uno alla volta

import random # il modulo random contiene classi e funzioni che ci permettono di creare oggetti in modo, appunto, random

randomlist = random.sample(range(-9000000, 9000000), 10000)
randomlist
```

```python
%%timeit
#1 li conservo in una lista appendendoli a sinistra o a destra a seconda del segno
lst = []
for i in randomlist:
    if i >= 0:
        lst.append(i)
    else:
        lst = [i] + lst
```

```python
# PROBLEMA: ricorsione con memoing
# trovare quanti numeri minori di n e maggiori di m possono essere fattorizzati solo con 5 e 3
m = 500
n = 3000
```

```python
%%timeit
mynums = Counter() # questa variabile è globale e rimane accessibile ed aggiornabile anche dall'interno di funzioni

def recursive_memo(i):
    if i < n and not mynums[i]:
        if i > m:
            mynums[i] += 1
        recursive_memo(i*3)
        recursive_memo(i*5)

recursive_memo(1)
```

```python
mynums
```

```python
%%timeit
mynums = set() 

def recursive_memo(i):
    if i < n and i not in mynums:
        if i > m:
            mynums.add(i)
        recursive_memo(i*3)
        recursive_memo(i*5)

recursive_memo(1)
```

```python
# non sempre le classi appartenenti a collections sono migliori di quelle built-in. Dipende sempre dal tipo di problema.
# le due soluzioni sono perfettamente analoghe a meno del tempo computazionale.
```

```python
# ESEMPIO da Hackerrank
# https://www.hackerrank.com/challenges/angry-children/problem?h_l=interview&playlist_slugs%5B%5D=interview-preparation-kit&playlist_slugs%5B%5D=greedy-algorithms
def maxMin(k, arr):
    arr.sort()
    n = len(arr)
    possdiff = 100000000000
    for i in range(n - k +1):
        if arr[i+k-1]-arr[i] < possdiff:
            possdiff = arr[i+k-1]-arr[i]
    return possdiff
```
