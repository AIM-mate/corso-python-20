````python
import requests as req
import bs4 as bs
````

````python
class AdelphiScraper():
  def __init__(self):
    self.catalog      = 'https://www.adelphi.it/catalogo/'
    self.subject      = 'https://www.adelphi.it/catalogo/materia/'
    self.subjects_map = self.init_subjects()

  def init_subjects(self):
    resp  = req.get(self.catalog)
    soup  = bs.BeautifulSoup(resp.text, "html.parser")
    items = soup.find_all('ul', class_='subject')
    ret = dict()
    for i in items:
      for j in i:
        ret.update({int(j.a['href'].split('/')[-1]) : j.text})
    return ret

  def get_subject_list(self):
    return self.subjects_map
  
  def get_subject_books(self, subject_code):
    page_found = True
    suffix = ''
    page_counter = 1
    books = dict()
    isbns = dict()
    while page_found:
      try:
        resp = req.get(self.subject + str(subject_code) + suffix, allow_redirects = False)
        soup = bs.BeautifulSoup(resp.text, "html.parser")
        if not len(soup):
          raise ValueError()
        book_infos, isbn_maps = self.extract_books(soup)
        books.update(book_infos)
        isbns.update(isbn_maps)
        page_counter += 1
        suffix = '/p' + str(page_counter)
      except:
        page_found = False
    return books, isbns

  def extract_books(self, soup):
    items = soup.find_all('div', class_='list_impressum')
    items2 = soup.find_all('div', class_='data' )
    items3 = soup.find_all('div', class_ = "abstract hidden-xs")
    isbn_map  = dict()
    book_info = dict()
    counter = 0
    for i in items:
      isbn_map.update({i.a['href'].split('/')[-1] : i.a.text}) #isbn : titolo
      book_info.update({i.a.text : {'Autore': i.div.text, \
                                    'Anno': items2[counter].text.split(' ')[0].split(',')[0], \
                                    'Prezzo Originale' : i.span.text.split(' ')[1].replace(',','.'), \
                                    'Sconto': i.span.text.split(' ')[2], \
                                    'Pagine' :items2[counter].text.split(' ')[2].split(',')[0].split('-')[-1], \
                                    'Abstract': items3[counter].text}})
      counter += 1
    return book_info, isbn_map
````

````python
def merge_two_dicts(x, y):
    z = x.copy()
    z.update(y)
    return z

Ad_Scraper = AdelphiScraper()
library = dict()
library_info = dict()
categories = [Ad_Scraper.get_subject_list()[i] for i in range(1, len(Ad_Scraper.get_subject_list()) + 1)]
for i in range(len(categories)):
    current = Ad_Scraper.get_subject_books(i)[0]
    current_category = categories[i]
    for j in current.keys():
        current[j]['Categoria'] = current_category
    merge_two_dicts(library_info, current)
    merge_two_dicts(library, Ad_Scraper.get_subject_books(i)[1])
````

````python
class Book():
    def __init__(self, ISBN):
        global library
        global library_info
        self.ISBN = ISBN
        self.title = library[ISBN]
        self.print_year = library_info[self.title]['Anno']
        self.author = library_info[self.title]['Autore']
        self.abstract = library_info[self.title]['Abstract']
        self.pages = int(library_info[self.title]['Pagine'])
        self.category = library_info[self.title]['Categoria']
        
    def get_whole_price(self):
        return float(library_info[self.title]['Prezzo Originale'])
    
    def get_net_price(self):
        return float(library_info[self.title]['Prezzo Originale'])*(1-int(library_info[self.title]['Sconto'][1:-1])/100)
    
    def print_book_infos(self):
        print(self.abstract)
        
    def abstract_len(self):
        print(len(self.abstract))
        
    def is_near(self, book_2)
        if self.category == book_2.category or (self.ISBN - book_2.ISBN) % 3 == 0:
            return True
        return False
````

````python
class MyLibrary():
    def __init__(self, ISBN_list):
        self.book_list = ISBN_list
        self.books = dict()
        for i in range len(ISBN_list):
            current_ISBN = ISBN_list[i]
            self.books[current_ISBN] = Book(current_ISBN)
        
    def insert_book(self, ISBN):
        if ISBN not in self.book_list:
            self.books[ISBN] = Book(ISBN)
            self.book_list.append(ISBN)
        
    def remove_book(self, ISBN):
        if ISBN in self.book_list:
            del self.books[ISBN]
            self.book_list.remove(ISBN)
        else:
            print("Libro non trovato")
    
    def get_library_value(self):
        value = 0
        net_value = 0
        for ISBN in self.book_list:
            value += self.books[ISBN].get_whole_price()
            net_value += self.books[ISBN].get_net_price()
        return value, net_value
    
    def get_total_pages(self):
        pages = 0
        for ISBN in self.book_list:
            pages += self.books[ISBN].pages
        return pages

    
    def minimum_path(self, start_ISBN, end_ISBN):
        def min_path_rec(book_list, current_list, end_ISBN, cost):
            current_cost = -1
            for ISBN in book_list:
                if ISBN not in current_list and current_list[-1].is_near(ISBN):
                    if ISBN == end_ISBN:
                        current_list.append(ISBN)
                        return current_list, cost+1
                    temp_list = current_list
                    temp_list.append(ISBN)
                    temp_list, temp_cost = min_path_rec(book_list, temp_list, end_ISBN, cost+1)
                    if temp_cost >= 0 and (current_cost > temp_cost or current_cost == -1):
                        next_list = temp_list
                        current_cost = temp_cost
            if current_cost >= 0:
                return next_list, current_cost
            else:
                return [], -1
        
        
        if start_ISBN not in self.book_list or end_ISBN not in self.book_list:
            return []
        return min_path_rec(self.book_list, [start_ISBN], end_ISBN, 0)
    
    def minimum_cost_path(self, start_ISBN, end_ISBN):
        def min_cost_path_rec(book_list, current_list, end_ISBN, cost, books):
            current_cost = -1
            for ISBN in book_list:
                if ISBN not in current_list and current_list[-1].is_near(ISBN):
                    if ISBN == end_ISBN:
                        current_list.append(ISBN)
                        return current_list, cost+books[ISBN].get_net_price()
                    temp_list = current_list
                    temp_list.append(ISBN)
                    temp_list, temp_cost = min_cost_path_rec(book_list, temp_list, end_ISBN, cost+books[ISBN].get_net_price(), books)
                    if temp_cost >= 0 and (current_cost > temp_cost or current_cost == -1):
                        next_list = temp_list
                        current_cost = temp_cost
            if current_cost >= 0:
                return next_list, current_cost
            else:
                return [], -1
        
        
        if start_ISBN not in self.book_list or end_ISBN not in self.book_list:
            return []
        return min_path_rec(self.book_list, [start_ISBN], end_ISBN, 0.0, self.books)
````
