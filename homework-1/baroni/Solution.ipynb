```python
Stocks = [ENEL,ENI,RACE,ISP] = [[stocks_today[i] for stocks_today in Stocks_Value] for i in range(0,4)]
print("\n Ordered Stocks:")
_ = [print(st) for st in Stocks]
```

```python
num_days = len(Stocks[0])

Stocks_Mean = [sum(Stocks[i])/num_days for i in range(0,4)]
print("\n Means:")
print(Stocks_Mean)

Stocks_Std = [((sum([(so-Stocks_Mean[i])**2 for so in Stocks[i]])/(num_days-1)))**0.5 for i in range(0,4)]
print("\n Std:")
print(Stocks_Std)

Correlations = [[sum([(Stocks[i][k]-Stocks_Mean[i])*(Stocks[j][k]-Stocks_Mean[i]) for k in range(0,num_days)])
                 /(num_days-1)/Stocks_Std[i]/Stocks_Std[j] for i in range(0,4)] for j in range(0,4)]

print("\n Correlations:")
_ = [print(corr) for corr in Correlations]
```
