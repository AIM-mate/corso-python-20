```python
# 1. Trasponi la lista di liste delle azioni (invertendo righe e colonne) - 1 riga di codice

Stocks = [ENEL,ENI,RACE,ISP] = [list(s) for s in zip(*Stocks_Value)]

# Alternativa:
# Stocks = [ENEL,ENI,RACE,ISP] = [[stocks_today[i] for stocks_today in Stocks_Value] for i in range(0,4)]

print("\n Ordered Stocks:")
_ = [print(st) for st in Stocks]
```

```python
# 2. Trova i valori attesi (media) dei prezzi di ognuna delle 4 azioni - 1 riga di codice

Stocks_Mean = [sum(st)/len(st) for st in Stocks]

# Alternativa:
# Stocks_Mean = [sum(Stocks[i])/len(Stocks[0]) for i in range(len(Stocks))]

print("\n Means:")
print(Stocks_Mean)

# 3. Trova le deviazioni standard dei prezzi di ognuna delle 4 azioni - 1 riga di codice

Stocks_Std = [(sum([(st - stm[1])**2 for st in stm[0]])/(len(stm[0])-1))**0.5 for stm in zip(Stocks,Stocks_Mean)]

# Alternativa
# Stocks_Std = [(sum([(so-Stocks_Mean[i])**2 for so in Stocks[i]])/(len(Stocks[0])-1))**0.5 for i in range(len(Stocks))]

#Stocks_Std = [)(sum([(sv-Stocks_Mean[0])**2 for sv in Stocks[0]])/(len(Stocks[0])-1)))**0.5] #for st in zip(Stocks,Stocks_Mean)]

print("\n Std:")
print(Stocks_Std)

# 4. Salva in una lista di liste tutte le possibili correlazioni tra i prezzi di ognuna delle 4 azioni - 1 riga di codice

Correlations = [[sum([(Stocks[i][k]-Stocks_Mean[i])*(Stocks[j][k]-Stocks_Mean[i]) for k in range(len(Stocks[0]))])/(len(Stocks[0])-1)/Stocks_Std[i]/Stocks_Std[j] for i in range(len(Stocks))] for j in range(len(Stocks))]

# Alternativa:
# Correlations = [[sum([(st[0]-st1[1])*(st[1]-st2[1]) for st in zip(st1[0],st2[0])])/(len(st1[0])-1)/st1[2]/st2[2] for st1 in zip(Stocks,Stocks_Mean,Stocks_Std)] for st2 in zip(Stocks,Stocks_Mean,Stocks_Std)]

print("\n Correlations:")
_ = [print(corr) for corr in Correlations]
```
