```python
#Random walk 2D
n = 1000

x = [0]
y = [0]

direction = ["N", "S", "E", "O"]


try:
    for i in range(0,n-1):
        inp = input("Inserisci una direzione: \n")
        d = str(inp)
        
        if d == "N":
            step_x = x[i]
            step_y = y[i] + 1
        elif d == "E":
            step_x = x[i] - 1
            step_y = y[i] 
        elif d == "S":
            step_x = x[i]
            step_y = y[i] - 1
        elif d == "O":
            step_x = x[i] + 1
            step_y = y[i]
        x.append(step_x)
        y.append(step_y)
    print(x)
    print(y)

except:
    print("Inserisci una direzione")
```
