```python
####PART 2
def LogisticMap (x):
    r=3.7
    ##your code goes here
    y = r*x*(1-x)
    return(y)
```

```python
###PART 3
def Cryptate (xi,x):
    b=True
    epsilon=1/256
    iterazioni=0
    while b==True:
        ##your code goes here
        c = xi/epsilon
        if (int(c) == ord(x)):
            return iterazioni,xi;
        ##your code goes here
        else:
            iterazioni += 1
            xi = LogisticMap(xi)
```
