```python
####PART 2
def LogisticMap (x):
    r = 3.7
    y = r*x*(1-x)
    return(y)
```

```python
###PART 3
def Cryptate (xi,x):
    b=True
    epsilon=1/256
    iterazioni=0
    while b==True:
        c = xi/epsilon
        if(int(c)==ord(x)):
            return iterazioni,xi;
        iterazioni += 1
        xi = LogisticMap(xi)
```
