**Utilità di Python per l'interfaccia utente e gestione degli errori**

1.   La funzione nativa di Python *input()* permette di assegnare un valore ad una variabile in base al valore che inseriremo dalla tastiera. L'idea è di usarla per comunicare le mosse al programma.
2.   Si dovrà prestare grande attenzione agli errori che possono sorgere da una mossa non ammissibile: non è consentito ad esempio muoversi fuori dalla scacchiera, inserire valori non numerici come lunghezza della mossa e molte altre eccezioni. Python è dotato del blocco *try-except* che ci consente di verificare se una azione procurerebbe un errore e di fornire una alternativa di esecuzione in caso.

Segue un esempio di utilizzo.

```python
print("Insert a number between 0 and 5!")

x = input() # subito dopo aver fatto girare questa cella di codice si dovrà inserire un valore dalla tastiera nell'apposito spazio che si creerà sotto
x_type = type(x) # la funzione type() è cruciale per la gestione di errori

print(x_type)
x = int(x) # attenzione: l'input da tastiera è sempre letto come stringa di default, si avrà bisogno di convertirlo ad intero

try: # adesso il codice proverà ad eseguire la parte di codice indentata sotto a try, se nel farlo incontrerà degli errori deciderà invece di eseguire quella sotto ad except
  if x not in [0,1,2,3,4,5]: # verifichiamo che x sia compreso tra 0 e 5
    raise ValueError()  # questa linea di codice serve a far generare manualmente un errore al codice
  else:
    print("The double of your number is {}".format(2*x))     
except:
  print("I said between 0 and 5!") # qualora x non fosse un numero tra 0 e 5, il precedente blocco di codice genererebbe un errore e giungerebbe qui


# Hint: accedere ad una struttura dati in una sua porzione che non esiste è un errore!

triangle = [[1],[2,2],[3,4,3]] 
#print(triangle[1][1])
coords = (2,1)

try:
  print(triangle[coords[0]][coords[1]])
except:
  print("Those coordinates are out of boundaries")
```

**Regole**

1.   Per ogni mossa, sarà richiesto all'utente di inserire due variabili dalla tastiera: la prima sarà una lettera tra L, R, U, D ed indicherà la direzione dello spostamento, la seconda sarà un intero ed indicherà la quantità di caselle del movimento. Nessun altro tipo di input è tollerato e si dovrà notificare all'utente la scritta "Mossa non riconosciuta, prego ripetere" e permettergli di provare ad inserire di nuovo la mossa senza perdere i progressi fatti.
2.   Qualora lo spostamento fosse accettabile ma dovesse condurre al di fuori della scacchiera, si notifichi all'utente con il messaggio "Questo spostamento non è consentito, provarne un altro" e si permetta di inserire un altra mossa al suo posto senza perdere i progressi fatti.
3.   Qualora il giocatore voglia smettere di giocare si lasci che inserisca la lettera E e glielo si comunichi "Grazie di aver partecipato!"
*Bonus*
3.   Si comunichi all'utente la sua posizione iniziale definita nella prima cella di codice e si lasci che prima di iniziare il gioco scelga una posizione finale da raggiungere, qualora dovesse farcela ci si congratuli con lui con il messaggio "Complimenti, hai raggiunto l'obiettivo!".

```python
import random
size = 2
board =[[0 for i in range(size)] for j in range(size)]
x=0
y=0
board[y][x]=1
stat2 = 1
xwish=random.randrange(size)
ywish=random.randrange(size)
print("Devi andare in:")
print(xwish,ywish)
while(stat2==1):
    
    print(board)
    print("Scegliere una direzione tra sinistra (L), destra (R), in alto (U) o in basso (D). Scrivi (E) per uscire.")
    direction=input()
    direction_type = type(direction) 
    
    try: 
        if direction not in ["L","R","U","D","E"]:
            raise ValueError()  
        else:
            if direction=="E":
                print("Grazie di aver partecipato! :( Ma sappi che solo i deboli si arrendono...")
                stat2=0
            else:
                print("Scegliere una distanza da percorrere")
                length=input()
                length=int(length) 
                print(length)
                try: 
                    if direction=="L" and x-length>=0:
                        board[y][x]=0
                        x=x-length
                        board[y][x]=1
                        
                    elif direction=="R" and x+length<size:
                        board[y][x]=0
                        x=x+length
                        board[y][x]=1
                       
                    elif direction=="U" and y-length>=0:
                        board[y][x]=0
                        y = y-length
                        board[y][x]=1
    
                    elif direction=="D" and y+length<size:
                        board[y][x]=0
                        y = y+length
                        board[y][x]=1
                        
                    else:
                        raise ValueError()
                    
                    if x==xwish and y==ywish:
                        print("Congratulazioni! Sei arrivato a destinazione!")
                        stat2=0
                        
                except: 
                    print("Stai uscendo dalla tabella. Riprova!")
    except: 
        print("Direzione non valida. Riprova!")
    
```
