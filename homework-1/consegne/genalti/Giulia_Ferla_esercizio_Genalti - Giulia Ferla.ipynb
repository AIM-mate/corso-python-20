```python
# dimensione scacchiera
nx = 3
ny = 3

# scacchiera
scac = [[0] * ny for i in range(nx)]
currentx = int(input("Inserisci coordinata x iniziale: "))
currenty = int(input("Inserisci coordinata y iniziale: "))
finalx = int(input("Inserisci coordinata x finale: "))
finaly = int(input("Inserisci coordinata y finale: "))
scac[currentx][currenty] = 1
scac[finalx][finaly] = "F"
for i in scac:
    print(i)


while(True):
    direction = input("inserisci la direzione di spostamento: L, R, U, D: ")
    if direction == "E":
        print("Grazie di aver partecipato!")
        break
    length = int(input("Inserisci lo spostamento: "))
    repeat = False
    try:
        if direction not in ["L", "R", "U", "D"] or length <= 0:
            raise ValueError()
    except:
        repeat = True
        print("Mossa non riconosciuta, prego ripetere")

    if not repeat:
        if direction in ["L", "U"]:
            minus = -1
        else:
            minus = 1
    
        try:
            if direction in ["L", "R"]:
                newy = currenty + minus * length
                if newy == -1:
                    raise ValueError()
                scac[currentx][newy] = 1
                scac[currentx][currenty] = 0
                currenty = newy
            else:
                newx = currentx + minus * length
                if newx == -1:
                    raise ValueError()
                scac[newx][currenty] = 1
                scac[currentx][currenty] = 0
                currentx = newx
        except:
            print("Questo spostamento non è consentito, provarne un altro")
        
    
    if currentx == finalx and currenty == finaly:
        print("Complimenti, hai raggiunto l'obiettivo!")
        break

    for i in scac:
        print(i)
    print("Prossima mossa: ")
```

```python
# problema inverso

# dimensione scacchiera
nx = 3
ny = 3

# scacchiera
scac = [[0] * ny for i in range(nx)]
currentx = int(input("Inserisci coordinata x iniziale: "))
currenty = int(input("Inserisci coordinata y iniziale: "))
finalx = int(input("Inserisci coordinata x finale: "))
finaly = int(input("Inserisci coordinata y finale: "))
scac[currentx][currenty] = 1
scac[finalx][finaly] = "F"
for i in scac:
    print(i)
print("\n")
spostamento = ["N", None]

while True:
    
    if currentx != finalx:
        d = finalx-currentx
        if d < 0:
            spostamento[0] = "U"
            spostamento[1] = d * (-1)
        else:
            spostamento[0] = "D"
            spostamento[1] = d
        scac[currentx][currenty] = 0
        currentx = finalx
        scac[currentx][currenty] = 1
            
    else:
        if currenty != finaly:
            d = finaly-currenty
            if d < 0:
                spostamento[0] = "L"
                spostamento[1] = d * (-1)
            else:
                spostamento[0] = "R"
                spostamento[1] = d
            scac[currentx][currenty] = 0
            currenty = finaly
            scac[currentx][currenty] = 1
        
    print("Mossa:", spostamento[0], spostamento[1])
    for i in scac:
        print(i)
    
    if currentx == finalx and currenty == finaly:
        print("Obiettivo raggiunto!")
        break
```
