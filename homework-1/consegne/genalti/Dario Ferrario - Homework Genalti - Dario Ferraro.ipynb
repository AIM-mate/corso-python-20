```python
"""Gioco in cui √® il giocatore a scegliere come muoversi"""

n = eval(input("Righe della scacchiera: "))
m = eval(input("Colonne della scacchiera: "))
board = []
for i in range(n): #Creazione scacchiera
    board.append(['‚òê'])
    for j in range(m):        
        board[i].extend(['‚òê'])

posizione = [0,0] #Coordinate iniziali
board[posizione[0]][posizione[1]] = 'üêà' #Posiziono il giocatore nelle coordinate iniziali
for i in range(n): #Stampo la prima scacchiera
    for j in range(n):                
        print(board[i][j], end="\t")

    print()

print("Qual √® la tua meta (scrivere a,b con a <",n," e b <",m,")?\n")
meta = tuple(eval(input()))
while meta[0] >= n or meta[1] >= m: #Verifico che la meta sia all'interno della scacchiera
    if meta[0] >= n or meta[1] >= m:        
        print("La meta eccede i limiti della scacchiera")
        meta = tuple(eval(input("Qual √® la tua meta? ")))

board[meta[0]][meta[1]] = 'üêÅ' #Posiziono la meta
for i in range(n): #Stampo la scacchiera
    for j in range(n):
        print(board[i][j], end="\t")
    
    print()
            
finito = False 
while finito == False: #Il gioco continua finch√© finito = True
    direction = input("Direzione dello spostamento (L = left, R = right, U = up, D = down): ")
    dir_corretta = False
    while dir_corretta == False:  #Verifico che la direzione inserita sia una lettera richiesta    
        try:
            if direction not in ['L','D','R','U','l','r','u','d']:                
                raise ValueError()
            
            else:                
                dir_corretta = True #Cambia la variab
        
        except:            
            print("La direzione deve essere L, R, D o U")
            direction = input("Direzione dello spostamento: ")

    length = eval(input("Di quante caselle √® lo spostamento: "))
    len_out = False
    while len_out == False: #Verifico che lo spostamento non esca dai limiti della scacchiera    
        try:            
            if direction in ['L','l'] and (posizione[1] - length) < 0:                
                raise ValueError()
            
            elif direction in ['R','r'] and (posizione[1] + length) >= m:                
                raise ValueError()
            
            elif direction in ['U','u'] and (posizione[0] - length) < 0:                
                raise ValueError()
            
            elif direction in ['D','d'] and (posizione[0] + length) >= n:                
                raise ValueError()
            
            else:                
                len_out = True
            
        except:             
            print("Lo spostamento eccede i limiti della scacchiera") 
            length = eval(input("Di quante caselle √® lo spostamento: "))
    
    board[posizione[0]][posizione[1]] = 'üêæ' #Lascio traccia di dove sono gi√† passato
    if direction in ['L','l']: #Modifico la posizione del giocatore
        posizione[1] = posizione[1] - length
                
    elif direction in ['R','r']:        
        posizione[1] = posizione[1] + length
            
    elif direction in ['U','u']:        
        posizione[0] = posizione[0] - length
            
    else:    
        posizione[0] = posizione[0] + length
    
    board[posizione[0]][posizione[1]] = 'üêà' #Posiziono il giocatore nelle nuove coordinate       
    for i in range(n): #Stampo la scacchiera
        for j in range(n):
            print(board[i][j], end="\t")
                
        print()
        
    if posizione[0] == meta[0] and posizione[1] == meta[1]: #Se raggiungo la meta il gioco finisce       
        finito = True
        print("Hai raggiunto la meta!")
        
    else:            
        domanda = input(" Vuoi continuare (N = no)? ") #Se il giocatore non vuole pi√π giocare pu√≤ interrompere
        if domanda in ['No','no','n','N']:        
            finito = True                 

print("Game over") #Fine del ciclo con finito e fine del gioco
```

```python
"""Gioco in cui √® il computer a scegliere come muoversi"""
n = eval(input("Righe della scacchiera: "))
m = eval(input("Colonne della scacchiera: "))
board = []
for i in range(n): #Creazione scacchiera
    board.append(['‚òê'])
    for j in range(m):        
        board[i].extend(['‚òê'])

a,b = eval(input("Posizione iniziale (scrivere a,b): "))
posizione = [a,b]
board[posizione[0]][posizione[1]] = 'üêà' #Posiziono il giocatore nelle coordinate iniziali
for i in range(n): #Stampo la prima scacchiera
    for j in range(n):                
        print(board[i][j], end="\t")

    print()

print("Qual √® la meta (scrivere a,b con a <",n," e b <",m,")?\n")
meta = tuple(eval(input()))
while meta[0] >= n or meta[1] >= m: #Verifico che la meta sia all'interno della scacchiera
    if meta[0] >= n or meta[1] >= m:        
        print("La meta eccede i limiti della scacchiera")
        meta = tuple(eval(input("Qual √® la tua meta? ")))

if posizione[0] == meta[0] and posizione[1] == meta[1]: #Caso in cui la meta combaci con la posizione iniziale
    board[meta[0]][meta[1]] = 'üê±'
    for i in range(n): #Stampo la scacchiera
        for j in range(n):
            print(board[i][j], end="\t")
    
        print()
    
    print("La posizione iniziale combacia con la meta")
    
else:
    board[meta[0]][meta[1]] = 'üêÅ' #Posiziono la meta
    for i in range(n): #Stampo la scacchiera
        for j in range(n):
            print(board[i][j], end="\t")

        print()

    distanza_vert = posizione[0] - meta[0] #Distanza verticale tra posizione e meta
    distanza_oriz = posizione[1] - meta[1] #Distanza orizzontale tra posizione e meta
    if distanza_vert > 0: #Definisco la migliore mossa verticale per arrivare in meta
        mossa_vert = ['Su',abs(distanza_vert)]

    elif distanza_vert < 0:
        mossa_vert = ['Gi√π',abs(distanza_vert)]

    else:
        mossa_vert = 'Nessuna'

    if distanza_oriz > 0: #Definisco la migliore mossa orizzontale per arrivare in meta
        mossa_oriz = ['Sinistra',abs(distanza_oriz)]

    elif distanza_oriz < 0:
        mossa_oriz = ['Destra',abs(distanza_oriz)]

    else:
        mossa_oriz = 'Nessuna'

    if mossa_oriz != 'Nessuna' and mossa_oriz != 'Nessuna': #Stampo le mosse da fare
        print("La prima mossa della pedina √®:",mossa_oriz[0],"di",mossa_oriz[1],"casella/e")
        print("La seconda mossa della pedina √®:",mossa_vert[0],"di",mossa_vert[1],"casella/e")

    elif mossa_oriz == 'Nessuna':
        print("La mossa della pedina √®:",mossa_vert[0],"di",mossa_vert[1],"casella/e")

    elif mossa_vert == 'Nessuna':
        print("La mossa della pedina √®:",mossa_oriz[0],"di",mossa_oriz[1],"casella/e")
```
