L'obiettivo di questo esercizio è creare un programma che ci permetta di muovere un oggetto su una scacchiera. Il programma dovrà interfacciarsi con l'utente ricevendo in input da tastiera la direzione del movimento e la sua lunghezza ed aggiornare la struttura di conseguenza.

**Implementazione della scacchiera**

1.   Scegliere la struttura dati adeguata a rappresentare una scacchiera.
2.   Capire come rappresentare quella struttura dati in Python ed implementarla. Si provino varie dimensioni della scacchiera: 2x2, 10x8, 1000x1000...
3.   Scegliere un punto di partenza, quindi modificare la scacchiera in modo da "inserire" l'oggetto in quella determinata posizione.


```python
n_righe=int(input("Inserisci numero righe scacchiera: "))
n_colonne=int(input("Inserisci numero colonne scacchiera: "))
board = list()
for i in range(n_righe):
    riga=list()
    for j in range(n_colonne):
        riga.append((i,j))
    board.append(riga)
print(board)
initial_coords = (1,2)
board[initial_coords[0]][initial_coords[1]]="Pedina"
print("\nLa scacchiera è: ",board)
```

**Utilità di Python per l'interfaccia utente e gestione degli errori**

1.   La funzione nativa di Python *input()* permette di assegnare un valore ad una variabile in base al valore che inseriremo dalla tastiera. L'idea è di usarla per comunicare le mosse al programma.
2.   Si dovrà prestare grande attenzione agli errori che possono sorgere da una mossa non ammissibile: non è consentito ad esempio muoversi fuori dalla scacchiera, inserire valori non numerici come lunghezza della mossa e molte altre eccezioni. Python è dotato del blocco *try-except* che ci consente di verificare se una azione procurerebbe un errore e di fornire una alternativa di esecuzione in caso.

Segue un esempio di utilizzo.

**Regole**

1.   Per ogni mossa, sarà richiesto all'utente di inserire due variabili dalla tastiera: la prima sarà una lettera tra L, R, U, D ed indicherà la direzione dello spostamento, la seconda sarà un intero ed indicherà la quantità di caselle del movimento. Nessun altro tipo di input è tollerato e si dovrà notificare all'utente la scritta "Mossa non riconosciuta, prego ripetere" e permettergli di provare ad inserire di nuovo la mossa senza perdere i progressi fatti.
2.   Qualora lo spostamento fosse accettabile ma dovesse condurre al di fuori della scacchiera, si notifichi all'utente con il messaggio "Questo spostamento non è consentito, provarne un altro" e si permetta di inserire un altra mossa al suo posto senza perdere i progressi fatti.
3.   Qualora il giocatore voglia smettere di giocare si lasci che inserisca la lettera E e glielo si comunichi "Grazie di aver partecipato!"
*Bonus*
3.   Si comunichi all'utente la sua posizione iniziale definita nella prima cella di codice e si lasci che prima di iniziare il gioco scelga una posizione finale da raggiungere, qualora dovesse farcela ci si congratuli con lui con il messaggio "Complimenti, hai raggiunto l'obiettivo!".

```python
coord = initial_coords;
print("Posizione iniziale: ", coord)

coord_fin=list()
print("Posizione finale da raggiungere:")
print("Inserire coordinata x: ")
coord_fin.append(int(input()))
print("Inserire coordinata y: ")
coord_fin.append(int(input()))
coord_fin=tuple(coord_fin)
print("La posizione finale da raggiungere è: ",coord_fin)

while True:
    print("Inserisci la direzione di spostamento: L, R, U, D: ")
    direzione = input()
    lunghezza = input("Inserisci la quantità di caselle, numero intero positivo: ")
    
    try:
        lunghezza = int(lunghezza)
        if direzione == 'E':
            print("Grazie di aver partecipato")
            break
        if (direzione not in ['L', 'R', 'U', 'D']) or (lunghezza<=0):
            raise ValueError()
    except:
        print("Mossa non riconosciuta, prego ripetere")
        
    try:
        if direzione == 'L':
            if (coord[1] - lunghezza) < 0:
                raise ValueError()
            board[coord[0]][coord[1]] = coord
            board[coord[0]][coord[1] - lunghezza] ='Pedina'
            cambio_coord = list(coord)
            cambio_coord[1] = coord[1] - lunghezza
            coord = tuple(cambio_coord)
            print("La pedina è in posizione: ", coord)
            
        elif direzione == 'R':
            if (coord[1] + lunghezza) < 0:
                raise ValueError()
            board[coord[0]][coord[1]] = coord
            board[coord[0]][coord[1] + lunghezza] = 'Pedina'
            cambio_coord = list(coord)
            cambio_coord[1]=  coord[1] + lunghezza
            coord = tuple(cambio_coord)
            print("La pedina è in posizione: ", coord)
            
        elif direzione == 'U':
            if (coord[0] - lunghezza) < 0:
                raise ValueError()
            board[coord[0]][coord[1]] = coord
            board[coord[0] - lunghezza][coord[1]] = 'Pedina'
            cambio_coord = list(coord)
            cambio_coord[0] = coord[0] - lunghezza
            coord = tuple(cambio_coord)
            print("La pedina è in posizione: ", coord)
            
        else:
            if (coord[1] + lunghezza) < 0:
                raise ValueError()
            board[coord[0]][coord[1]] = coord
            board[coord[0] + lunghezza][coord[1]] = 'Pedina'
            cambio_coord = list(coord)
            cambio_coord[0] = coord[0] + lunghezza
            coord = tuple(cambio_coord)
            print("La pedina è in posizione: ", coord)
        print("Scacchiera aggiornata: ", board)
        if coord == coord_fin:
            print("Complimenti, hai raggiunto l'obiettivo!")
            break
    except:
        print("Questo spostamento non è consentito, provane un altro")
print("Done")
```

**Problema inverso**

Adesso sarà il programma stesso a muovere il pezzo. Si lasci decidere all'utente la coordinata della destinazione, dopodichè il programma dovrà trovare le mosse che lo conducano alla meta nel minor numero di caselle attraversate (In una scacchiera bidimensionale ci aspetteremo al più due mosse, chiaramente). Si stampi all'utente quali sono le mosse scelte.

```python
n_righe=int(input("Inserisci numero righe scacchiera: "))
n_colonne=int(input("Inserisci numero colonne scacchiera: "))
board = list()
for i in range(n_righe):
    riga=list()
    for j in range(n_colonne):
        riga.append((i,j))
    board.append(riga)
coord_iniz = (1,2)
print("Posizione iniziale: ", coord_iniz)
board[coord_iniz[0]][coord_iniz[1]]="Pedina"
print("La scacchiera è:\n",board)

print("Posizione finale da raggiungere:")
not_exit = True
while not_exit:
    coord_fin = list()
    print("Inserire coordinata x: ")
    x = input()
    print("Inserire coordinata y: ")
    y = input()
    try:
        x = int(x)
        coord_fin.append(x)
        y = int(y)
        coord_fin.append(y)
        if coord_fin[0] < 0 or coord_fin[0] >= n_righe:
            raise ValueError()
        if coord_fin[1] < 0 or coord_fin[1] >= n_colonne:
            raise ValueError()
        coord_fin=tuple(coord_fin)
        print("La posizione finale da raggiungere è: ",coord_fin)
        not_exit = False
    except: 
        print("Posizione non ammissibile, riprovare")

first_move = []
second_move = []

if coord_iniz == coord_fin:
    print("Non sono necessarie mosse per raggiungere la destinazione")
    
elif coord_iniz[0] != coord_fin[0] and coord_iniz[1] != coord_fin[1]:
    if coord_iniz[0] < coord_fin[0]:
        first_move.append('D')
        first_move.append(coord_fin[0] - coord_iniz[0])
    else: 
        first_move.append('U')
        first_move.append(coord_iniz[0] - coord_fin[0])
    if coord_iniz[1] < coord_fin[1]:
        second_move.append('R')
        second_move.append(coord_fin[1] - coord_iniz[1])
    else:
        second_move.append('L')
        second_move.append(coord_iniz[1] - coord_fin[1])
    print("Mosse da svolgere: ", first_move, second_move)

elif coord_iniz[0] == coord_fin[0]:
    if coord_iniz[1] < coord_fin[1]:
        first_move.append('R')
        first_move.append(coord_fin[1] - coord_iniz[1])
    else:
        first_move.append('L')
        first_move.append(coord_iniz[1] - coord_fin[1])
    print("Mossa da svolgere: ", first_move)

else:
    if coord_iniz[0] < coord_fin[0]:
        first_move.append('D')
        first_move.append(coord_fin[0] - coord_iniz[0])
    else:
        first_move.append('U')
        first_move.append(coord_iniz[0] - coord_fin[0])
    print("Mossa da svolgere: ", first_move)
print("Done")
```
