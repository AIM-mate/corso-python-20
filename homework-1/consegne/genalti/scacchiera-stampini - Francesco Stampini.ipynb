**Implementazione della scacchiera**

```python
i, j = 10, 8
board = [[0 for x in range(i)] for y in range(j)] 

initial_coord = [0, 0]
ic = initial_coord

board[ic[0]][ic[1]]=1
board
```

**Regole**

1.   Per ogni mossa, sarà richiesto all'utente di inserire due variabili dalla tastiera: la prima sarà una lettera tra L, R, U, D ed indicherà la direzione dello spostamento, la seconda sarà un intero ed indicherà la quantità di caselle del movimento. Nessun altro tipo di input è tollerato e si dovrà notificare all'utente la scritta "Mossa non riconosciuta, prego ripetere" e permettergli di provare ad inserire di nuovo la mossa senza perdere i progressi fatti.
2.   Qualora lo spostamento fosse accettabile ma dovesse condurre al di fuori della scacchiera, si notifichi all'utente con il messaggio "Questo spostamento non è consentito, provarne un altro" e si permetta di inserire un altra mossa al suo posto senza perdere i progressi fatti.
3.   Qualora il giocatore voglia smettere di giocare si lasci che inserisca la lettera E e glielo si comunichi "Grazie di aver partecipato!"
*Bonus*
3.   Si comunichi all'utente la sua posizione iniziale definita nella prima cella di codice e si lasci che prima di iniziare il gioco scelga una posizione finale da raggiungere, qualora dovesse farcela ci si congratuli con lui con il messaggio "Complimenti, hai raggiunto l'obiettivo!".

```python
# posizione
p = ic;
t = True;
def changep(p):
    board = [[0 for x in range(i)] for y in range(j)] 
    board[p[0]][p[1]]=1
    return(board)

while t == True:
    # fintanto che si vuole continuare (o che non si è raggiunto la meta)
    direction = input('Direzione:\n')
    length    = input('Lunghezza:\n')
      # si facciano le dovute verifiche sulla legittimità della mossa
    length=int(length)
    if direction== 'L' or direction == 'l':
        if p[1]-length>=0:
            p[1] = p[1]-length
            board=changep(p)
            print('La tua scacchiera: \n')
            for s in board:
                print(*s)
        else:
             print("Questo spostamento non è consentito, provarne un altro.\n")
    elif direction == 'R' or direction == 'r':
        if p[1]+length<i:
            p[1] = p[1]+length
            board=changep(p)
            print('La tua scacchiera: \n')
            for s in board:
                print(*s)
        else:
             print("Questo spostamento non è consentito, provarne un altro.\n")
    elif direction == 'U' or direction == 'u' :
        if p[0]-length>=0:
            p[0] = p[0]-length
            board=changep(p)
            print('La tua scacchiera: \n')
            for s in board:
                print(*s)
        else:
             print("Questo spostamento non è consentito, provarne un altro.\n")
    elif direction == 'D' or direction == 'd':
        if p[0]+length<j:
            p[0] = p[0]+length
            board=changep(p)
            print('La tua scacchiera: \n')
            for s in board:
                print(*s)
        else:
             print("Questo spostamento non è consentito, provarne un altro.\n")
    elif direction == 'E'or direction == 'e':
        print("Grazie di aver partecipato!")
        t=False;
    else:
        print("Inserire risposta valida.")
            # si aggiorni la posizione sulla scacchiera, Hint: si tenga traccia della posizione dell'utente costantemente 
```

**Problema inverso**

Adesso sarà il programma stesso a muovere il pezzo. Si lasci decidere all'utente la coordinata della destinazione, dopodichè il programma dovrà trovare le mosse che lo conducano alla meta nel minor numero di caselle attraversate (In una scacchiera bidimensionale ci aspetteremo al più due mosse, chiaramente). Si stampi all'utente quali sono le mosse scelte.

```python
x = input('coordinata x:\n')
y = input('coordinata y:\n')
x = int(x)
y = int(x)

if x>p[1] and x<j:
    print('R-', x-p[1])
elif x<p[1] and x>=0:
    print('L-', p[1]-x)
elif x==p[1]:
    print('')
else:
    print('errore')

if y>p[0] and y<i:
    print('D-', y-p[0])
elif y<p[0] and y>=0:
    print('U-',p[0]-y)
elif y==p[0]:
    print('')
else:
    print('errore')

p=[y,x]
board=changep([y,x])
for s in board:
    print(*s)
```
