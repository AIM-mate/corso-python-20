```python
dimensione_verticale = 4
dimensione_orizzontale = 4
scacchiera = dict()
#Inizializzo la scacchiera come un dizionario. Il pezzo occupato dall'untente ha come value 1, gli altri hanno come value 0
for i in range(dimensione_verticale):
    for j in range(dimensione_orizzontale):
        scacchiera[(i,j)] = 0
#Posiziono il player in una casella a scelta
posizione_iniziale = (0,0)
scacchiera[posizione_iniziale] = 1
```

```python
import random
#Comincia il gioco
posizione_attuale = posizione_iniziale
print('La tua posizione iniziale è:', posizione_iniziale,'\n')
obiettivo = (random.randint(0,dimensione_verticale - 1),random.randint(0,dimensione_orizzontale - 1))
print('Il tuo obiettivo è raggiungere:', obiettivo, '\n')
while True:
    direzione = input('In che direzione vuoi muoverti? (L,R,U,D)\nDigitare E per uscire.\n').upper()
    if direzione == 'E':
        break
    elif direzione not in ['L','R','D','U']:
        print('Mossa non riconosciuta, prego ripetere.\n')
        continue
    numero_caselle = input('Di quante caselle vuoi muoverti?\n')
    try:
        numero_caselle = int(numero_caselle)
    except:
        print('Mossa non riconosciuta, prego ripetere.\n')
        continue
    if direzione == 'L':
        nuova_posizione = (posizione_attuale[0], posizione_attuale[1] - numero_caselle)
    elif direzione == 'R':
        nuova_posizione = (posizione_attuale[0], posizione_attuale[1] + numero_caselle)
    elif direzione == 'U':
        nuova_posizione = (posizione_attuale[0] - numero_caselle, posizione_attuale[1])
    elif direzione == 'D':
        nuova_posizione = (posizione_attuale[0] + numero_caselle, posizione_attuale[1])
    if scacchiera.get(nuova_posizione,-1) == 0:
        scacchiera[posizione_attuale] = 0
        scacchiera[nuova_posizione] = 1
        posizione_attuale = nuova_posizione
    else:
        print('Questo spostamento non è consentito, provarne un altro\n')
        continue
    if posizione_attuale == obiettivo:
        print("Congratulazioni, hai raggiunto l'obiettivo!\n")
        break
print('Grazie di aver partecipato!\n')
```
