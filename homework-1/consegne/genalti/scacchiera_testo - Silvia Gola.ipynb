**Utilità di Python per l'interfaccia utente e gestione degli errori**

1.   La funzione nativa di Python *input()* permette di assegnare un valore ad una variabile in base al valore che inseriremo dalla tastiera. L'idea è di usarla per comunicare le mosse al programma.
2.   Si dovrà prestare grande attenzione agli errori che possono sorgere da una mossa non ammissibile: non è consentito ad esempio muoversi fuori dalla scacchiera, inserire valori non numerici come lunghezza della mossa e molte altre eccezioni. Python è dotato del blocco *try-except* che ci consente di verificare se una azione procurerebbe un errore e di fornire una alternativa di esecuzione in caso.

Segue un esempio di utilizzo.

```python
#coords=initial_coords
coords=[initial_coords[0],initial_coords[1]]
print(coords)
goal_coords=[0,0]
print("Inserisci riga obiettivo")
goal_coords[0]=int(input())
print("Inserisci colonna obiettivo")
goal_coords[1]=int(input())
print("goal coords:", goal_coords)
maxlen=max(goal_coords[0],goal_coords[1])

dir="L"
while (dir!="E"):
    print("Insert a direction: L, R, U or D, or type E to exit")
    dir = input() # subito dopo aver fatto girare questa cella di codice si dovrà inserire un valore dalla tastiera nell'apposito spazio che si creerà sotto
    #x_type = type(x) # la funzione type() è cruciale per la gestione di errori
    #print(x_type)
    #x = int(x) # attenzione: l'input da tastiera è sempre letto come stringa di default, si avrà bisogno di convertirlo ad intero
    try: # adesso il codice proverà ad eseguire la parte di codice indentata sotto a try, se nel farlo incontrerà degli errori deciderà invece di eseguire quella sotto ad except
        if dir not in ["L","R","U","D","E"]: # verifichiamo che x sia compreso tra 0 e 5
            raise ValueError()
            #print("I said between  a direction: L, R, U or D, or type E to exit")
        elif dir=="E":
            print("Program will exit soon")
        else:
            print("Your direction is ok, Insert a number between 0 and ", maxlen)

            x0 = input() # subito dopo aver fatto girare questa cella di codice si dovrà inserire un valore dalla tastiera nell'apposito spazio che si creerà sotto
            x = int(x0) # attenzione: l'input da tastiera è sempre letto come stringa di default, si avrà bisogno di convertirlo ad intero
            if x not in range(0,maxlen+1): # verifichiamo che x sia compreso tra 0 e 5
                print("I said between 0 and ", maxlen) # qualora x non fosse un numero tra 0 e 5, il precedente blocco di codice genererebbe un errore e giungerebbe qui  # questa linea di codice serve a far generare manualmente un errore al codice
            else:
                if dir=="L":
                    if (coords[1]-x<0):
                        print("Mossa non riconosciuta, prego ripetere (troppo a sinistra!)")
                    else:
                        coords[1]=coords[1]-x
                elif dir=="R":
                    if (coords[1]+x>=board[1]):
                        print("Mossa non riconosciuta, prego ripetere (troppo a destra!)")
                    else:
                        coords[1]=coords[1]+x
                elif dir=="U":
                    if (coords[0]-x<0):
                        print("Mossa non riconosciuta, prego ripetere (troppo in alto!)")
                    else:
                        coords[0]=coords[0]-x
                elif dir=="D":
                    if (coords[0]+x>=board[0]):
                        print("Mossa non riconosciuta, prego ripetere (troppo in basso!)")
                    else:
                        coords[0]=coords[0]+x
                print(coords)
                if(coords==goal_coords):
                    print("Complimenti, hai raggiunto l'obiettivo!")
    except:
          print("I said a direction: L, R, U or D, or type E to exit") # qualora x non fosse un numero tra 0 e 5, il precedente blocco di codice genererebbe un errore e giungerebbe qui
print("Program exit done")

print(coords)
```
