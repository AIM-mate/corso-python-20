```python
dr=int(input("Inserisci numero righe\n"))
dc=int(input("Inserisci numero colonne\n"))
board=list(range(dc*dr))
ind=0
for i in range(dr):
    for j in range(dc):
        board[ind]=[i,j]
        ind+=1
while True:
    initial_coords=[]
    r0=int(input("Inserisci riga iniziale.Sia righe che colonne partono da valore 0\n"))
    initial_coords.append(r0)
    i0=int(input("Inserisci colonna iniziale"))
    initial_coords.append(i0)
    if initial_coords in board:
        break
    else:
        print("Coordinate non accettabili")
print(initial_coords)
```

```python
b=True
while b:
    a=True
    while True:
        direction=input("Inserisci un direzione di movimento tra le seguenti: L,R,U,D. Digitare E per uscire\n")
        if direction in list("LRUDE"):
            break
        else:
            print("Scegli direzione consentita")
    if direction=='E':
        break
    c=True
    while c:
        length=input("Inserisci di quante caselle vuoi muoverti in tale direzione\n")
        try:
            if int(length)>0:
                c=False
                length=int(length)
        except:
            print("Carattere non corretto per lunghezze")
    
    if direction=='D' and [initial_coords[0]+length,initial_coords[1]] in board:            
        initial_coords=[initial_coords[0]+length,initial_coords[1]]
        print("TI trovi in posizione: ",initial_coords)
        a=False
    elif direction=='U' and [initial_coords[0]-length,initial_coords[1]] in board:
        initial_coords=[initial_coords[0]-length,initial_coords[1]]
        print("TI trovi in posizione: ",initial_coords)
        a=False
    elif direction=='R' and [initial_coords[0],initial_coords[1]+length] in board:
        initial_coords=[initial_coords[0],initial_coords[1]+length]
        print("TI trovi in posizione: ",initial_coords)
        a=False
    elif direction=='L' and [initial_coords[0],initial_coords[1]-length] in board:
        initial_coords=[initial_coords[0],initial_coords[1]-length]
        print("TI trovi in posizione: ",initial_coords)
        a=False
    if initial_coords==final_coords:
        print("Grandioso. Hai vinto")
        break
    elif a==True:
        print("Questo spostamento non è consentito, provarne un altro")
print("Grazie per aver partecipato")
    
    
        
```

```python
dr=int(input("Inserisci numero righe\n"))    #creazione di una generica scacchiera
dc=int(input("Inserisci numero colonne\n"))
board=list(range(dc*dr))
ind=0
for i in range(dr):
    for j in range(dc):
        board[ind]=[i,j]
        ind+=1
start=[0,0]
while True:
    final=[]
    r0=int(input("Inserisci riga a cui vuoi arrivare.Sia righe che colonne partono da valore 0\n"))
    final.append(r0)
    i0=int(input("Inserisci colonna a cui vuoi arrivare"))
    final.append(i0)
    if final in board:
        break
    else:
        print("Coordinate non accettabili")
move1=['D',abs(final[0]-start[0])]
move2=['R',abs(final[1]-start[1])]
if final[0]-start[0]<0:
    move1[0]='U'
if final[1]-start[1]<0:
    move2[0]='L'
print(move1,'\n',move2)
               
```
