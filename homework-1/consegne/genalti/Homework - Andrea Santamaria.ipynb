### Zacchei: 
1) Insertion sort

```python
# Function to do insertion sort 
def insertionSort(arr): 
  
    # Traverse through 1 to len(arr) 
    for i in range(1,len(arr)) : 
  
        key = arr[i]             #pick key element
  
        # Move elements of arr[0..i-1], that are greater than key, to one position ahead of their current position 
        j = i-1
        
        while j > -1 and key < arr[j] : 
                arr[j+1] = arr[j]             #move element one position ahead
                j -= 1             #update j
                
        arr[j+1] = key 
        
    return arr
  
# Driver code to test above 
arr = [1, 12, 11, 13, 5, 6, 1, 2] 
print(arr)
print ("Sorted array is:") 
print(insertionSort(arr))
print("GG")
#ggggggggggggg
```

2) Bubble sort

```python
def bubbleSort(arr): 
    n = len(arr) 
  
    # Traverse through all array elements 
    for i in range(n):
    
  
        # Last i elements are already in place 
        for j in range(n-i-1):
  
            # traverse the array from 0 to n-i-1 
            # Swap if the element found is greater 
            # than the next element 
            if arr[j+1] < arr[j]:
                arr[j+1], arr[j] = arr[j], arr[j+1]
            
            
    return arr
  
# Driver code to test above 
arr = [1, 90, 64, 34, 25, 12, 22, 11, 90] 

print(arr)
print ("Sorted array is:") 
print(bubbleSort(arr))
print("GG")
#ggggggggggggg
```

### Vitellio:
1) Successione di Fibonacci

```python
a = 10
def fibo(n):
    f0 = 1; f1 = 1
    print(f0); print(f1)
    while n>2: # perché devo fare n print ed i primi 2 li faccio per f0 ed f1
        f = f0 + f1
        print(f)
        f0, f1 = f1, f
        n-=1
    return 

fibo(a)

def fibo_no_print(n):
    f = [1,1]
    for i in range(2,n):
        f.append(f[i-2]+f[i-1])
    return f

b = fibo_no_print(a)
print(b)
```

2) Un'altra successione

```python
n = 15
def suc(n):
    f0 = 1; f1 = 7; f2 = 18
    print(f0); print(f1); print(f2)
    while n>3:
        f = f2 + f0 - f1 + 2
        print(f)
        f0, f1, f2 = f1, f2, f
        n-=1
    return

def suc_no_print(n):
    f = [1, 7, 18]
    for i in range(3, n):
        f.append(f[i-1] + f[i-3] - f[i-2] + 2)
    return f

n = 15
suc(n)
v = suc_no_print(n)
print(v)
```

3) Triangolo di Tartaglia

```python
n = 10
def tart(n):
    t = []
    old = []
    i = 0
    while(i<n):   # se conto come prima riga il solo [1] il codice è questo qui, altrimenti sarà while(i<=n)
        t.append(1)
        old.append(1)
        print(t)
        for j in range(i):
            t[j+1] = old[j] + old[j+1]
        for k in range(len(t)):
            old[k] = t[k]
        i += 1

    return

tart(n)
```

### Fedeli
1) Generatore lineare congruenziale

```python
state = 14 # stato interno del generatore (inizializzato a piacere) -> con stato uguale, si otterrà sempre la stessa sequenza
m = 73
def LCG():
  a = 5
  c = 2
  m = 73
  global state # per modificare all'interno di una funzione una variabile globale è necessario dichiararla globale, per ora non tocchiamo questo punto e non ce ne preoccupiamo
  state = (a * state + c) % m

  return state
```

```python
periodo = 0
find = 0
x = [state]
while find == 0 :
    var = LCG()
    if var in x:
        idx = x.index(var) # ricavo l'indice del valore doppio
        var = LCG() # estraggo un altro numero e confronto con l'idx successivo
        x.append(var)
        if x[idx+1] == var: #se il nuovo numero estratto è prorio uguale a quello dopo in idx ho la seq di 2 numeri
            find = 1
        else:
            periodo += 1 #era solo un numero estratto due volte
    else:
        x.append(var)
        periodo += 1

print(periodo, x)
```

```python
x = y = 0
n = 1000
x_list = []
y_list = []
# [0 , 0.25) mi muovo in basso di 1, [0.25, 0.5) mi muovo a sinistra di 1, [0.5, 0.75) mi muovo a destra di uno
for i in range(n):
    x_list.append(x)
    y_list.append(y)
    #print(x_list[i], y_list[i])
    pos = LCG()/m
    if pos < 0.5:
        if pos < 0.25:
            y -= 1
        else:
            x -= 1
    elif pos < 0.75:
        x += 1
    else:
        y += 1
```
