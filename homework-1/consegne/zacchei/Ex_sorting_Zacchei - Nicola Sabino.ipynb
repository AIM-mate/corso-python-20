```python
def insertionSort(arr): 
  
    # Traverse through 1 to len(arr) 
    for i in range(1,len(arr)) : 
  
        key=arr[i] #pick key element
              
        # Move elements of arr[0..i-1], that are 
        # greater than key, to one position ahead 
        # of their current position 
        j = i-1
        
        while j >=0 and key < arr[j]: 
                arr[j+1] = arr[j] #move element one position ahead
                j -= 1            #update j
                arr[j+1] = key              
        
    return arr
```

```python
arr = [12, 11, 13, 5, 6, 9.81] 
print(arr) 
print ("Sorted array is:") 
print(insertionSort(arr))
```

```python
def bubbleSort(arr): 
    n = len(arr) 
  
    # Traverse through all array elements 
    for i in range(n):
    
  
        # Last i elements are already in place 
        for j in range(n-1-i):
  
            # traverse the array from 0 to n-i-1 
            # Swap if the element found is greater 
            # than the next element 
            if arr[j] > arr[j+1] :                    #condition
                 arr[j], arr[j+1] = arr[j+1], arr[j]  #swap                   
                                
            
    return arr
```

```python
arr = [64, 34, 25, 12, 22, 11, 90, 3.1415926535] 

print(arr)
print ("Sorted array is:") 
print(bubbleSort(arr))
```

```python
def merge_sort(values): 
  
    out = []
    if(  len(values) == 1):
        return values;
    half  = int(len(values) / 2)
    lower = merge_sort(values[:half])
    upper = merge_sort(values[half:])
    lower_len = len(lower)
    upper_len = len(upper)
    i = 0
    j = 0
    while i != lower_len or j != upper_len:
        if ( i != lower_len and (j == upper_len or lower[i] < upper[j])):
            out.append(lower[i])
            i += 1
        else:
            out.append(upper[j])
            j += 1

    return out
```

```python
a = [12, 11, 11, 13, 5, 6,6, 7, 155, 2.7182818284] 
print("Given array is") 
print(a) 
  
a = merge_sort(a) 
  
# Print output 
print("Sorted array is : ") 
print(a) 
```
