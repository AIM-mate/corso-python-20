```python
# Python program for implementation of Insertion Sort 
  
# Function to do insertion sort 
def insertionSort(arr): 
  
    # Traverse through 1 to len(arr) 
    for i in range(1,len(arr)) : 
        key=arr[i] 
        j=i-1
        while j>=0 and arr[j]>key : 
            arr[j],arr[j+1]=arr[j+1],arr[j] 
            j=j-1   
        
    return arr
  
  
# Driver code to test above 
arr = [12, 11, 13, 5, 6] 
print(arr)
print ("Sorted array is:") 
print(insertionSort(arr))
```

```python
# Python program for implementation of  
# MergeSort  
def merge_sort(values):
    if len(values)==0 or len(values)==1:
        return values
    n=len(values)/2
    if int(n)==(n):
        l1=merge_sort(list(values[:int(n)]))
        l2=merge_sort(list(values[int(n):]))
    else:
        l1=merge_sort(values[:int(n)+1])
        l2=merge_sort(values[int(n)+1:])
    while len(l1)>0 or len(l2)>0:
        if len(l1)>0 and len(l2)>0:
            if l1[0]>l2[0]:
                values.pop(0)
                values.append(l2[0])
                l2.pop(0)
            else:
                values.pop(0)
                values.append(l1[0])
                l1.pop(0)
        elif len(l2)>0:
            for i in l2:
                values.pop(0)
                values.append(i)
                l2.pop(0)
        else:
            for i in l1:
                values.pop(0)
                values.append(i)
                l1.pop(0)
    return values

# Input list 
a = [14, 11, 13, 5, 6, 7] 
print("Given array is") 
print(a) 
  
a = merge_sort(a) 
  
# Print output 
print("Sorted array is : ") 
print(a) 
```
