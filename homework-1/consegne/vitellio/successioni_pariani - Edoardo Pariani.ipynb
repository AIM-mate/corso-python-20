# Serie di Fibonacci

```python
n=int(input("Inserisci numero termini: "))
while n<0:
    n=int(input("Inserisci numero termini: "))
#i primi due numeri
f1=1
f2=1
i=2
#Casistica n<2
if n==0:
    print("Done")
elif n==1:
    print(f2)
    print("Done")
#Successione di Fibonacci
else:
    print(f2)
    print(f1)
    while i<n:
        #Creo Fn
        fn=f1+f2
        print(fn)
        #scambio di variabili per rendere la somma dinamica nel tempo
        f2=f1
        f1=fn
        i+=1
    print("Done")
```

# Triangolo di Tartglia

```python
#metodo con i fattoriali

n=int(input("Inserisci numero righe: "))
while n<0:
    n=int(input("Inserisci numero righe: "))
#Funzione per non scrivere il fattoriale ogni volta
def fact(num):
    fattoriale=1
    while num!=0:
        fattoriale*=num
        num-=1
    return fattoriale
#Contatori per righe e indice di ogni riga

i=0
j=0
#Nested while loops: si basa sullo sviluppo del binomio di newton
while i<=n:
    row=[]
    while j<=i:
        
        x=(fact(i)/(fact(i-j)*fact(j)))
        row.append(int(x))
        j+=1
    print(row)
    j=0
    i+=1
print("Done")
```

```python
#Metodo senza fattoriali

#Chiedo il numero di righe
n=int(input("Inserisci numero righe: "))
while n<0:
    n=int(input("Inserisci numero righe: "))
#Contatori per righe e elementi   
i=2
j=1
#Casistica n=0 allora stampo solo la prima riga
if n==0:
    row=[1]
    print(row)
    print("Done")
#Casistica n=1 allora stampo le prime due 
elif n==1:
    row1=[1]
    row2=[1,1]
    print(row1)
    print(row2)
    print("Done")
#Si basa sulla somma delle celle delle righe precedenti tramite lo store delle singole righe in una lista di liste
#ogni nuovo valore viene inserito in una row generica creata nel while ogni volta e stampata, essa perÃ² viene anche
#immagazzinata in rows
else:
    row1=[1]
    row2=[1,1]
    print(row1)
    print(row2)
    rows=[row1, row2]
    while i<=n:
        row=[1]
        while j<i:
            row.append(rows[i-1][j-1]+rows[i-1][j])
            j+=1
        row.append(1)
        print(row)
        rows.append(row)
        i+=1
        j=1
    print("Done")
        
    
```
