# Successione di Fibonacci
Stampa i primi n numeri della successione di Fibonacci.

```python
n = 15
lista = []
for i in range(n):
    if i==0 or i==1:
        lista.append(1)
    else:
        numero = lista[i-1] + lista[i-2]
        lista.append(numero)
print(lista)
```

# Un'altra successione
Scrivere i primi n elementi ($a_n$) di una successione cosÃ¬ definita: 
- i primi tre elementi valgono 1, 7, 18.
- i successivi (i>=4) valgono 
$a_n = a_{n-1} + a_{n-3} - a_{n-2} + 2$


Esempio: 1 7 18 14 5 11 22 18 9 15 26 22 13 19 30 


```python
n = 15
lista = []
for i in range(n):
    if i==0:
        lista.append(1)
    elif i==1:
        lista.append(7)
    elif i==2:
        lista.append(18)
    else:
        numero = lista[i-1] + lista[i-3] - lista[i-2] + 2
        lista.append(numero)
print(lista)
    
    
```

# Triangolo di Tartaglia
Scrivere le prime n righe del triangolo di Tartaglia 

```python
n = 16
def fattoriale(n):
    if n==0:
        return 1
    else:
        return n * fattoriale(n-1)
for m in range(n):
    lista = []
    for k in range(m+1):
        lista.append(int(fattoriale(m)/(fattoriale(m-k)*fattoriale(k))))
    print(lista)
    
```
