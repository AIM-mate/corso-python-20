```python
# Fibonacci

num = input("Inserire il numero di elementi per la serie di Fibonacci: ")
try:
    n = int(num)
    n0 = 1
    print(n0, end = " ")
    n1 = 1
    print(n1, end = " ")
    for k in range(2,n):
        n2 = n0 + n1
        print(n2, end = " ")
        n0 = n1
        n1 = n2
except:
    print("La prossima volta inserisci un numero")
```

```python
# Successione 2

num = input("Inserire il numero di elementi per la serie: ")
try:
    n = int(num)
    n0 = 1
    print(n0, end = " ")
    n1 = 7
    print(n1, end = " ")
    n2 = 18
    print(n2, end = " ")
    for k in range(3,n):
        n3 = n2 + n0 - n1 +2
        print(n3, end = " ")
        n0 = n1
        n1 = n2
        n2 = n3
except:
    print("La prossima volta inserisci un numero")
```

```python
# Triangolo di Tartaglia

num = input("Inserire il numero di righe del Triangolo di Tartaglia: ")

try:
    n = int(num)
   
    v = [1,1]
    print(v)
 
    if n > 1:
        for k in range(2,n+1):
            v1 = [0]*(len(v)+1)
            v1[0] = 1
            v1[len(v)] = 1
            for i in range(1,len(v)):
                v1[i] = v[i-1] + v[i]
               
            v = v1
            print(v)
    
except:
    print("La prossima volta inserisci un numero")
```
