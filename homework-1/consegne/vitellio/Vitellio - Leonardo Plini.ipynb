```python
#esercizio vitellio
l=[ 1, 7,18]
n=0
while n<15:
    l.append(l[n]+l[n+2]-l[n+1]+2)
    n+=1
    
     
print(l)    
    
    
```

```python
l=[1]
print(l)
print('....')
n=0
while n<15:
    l.append(1)
    s=l[:]
    i=1
    while i<((n/2)+1):
        s[i]=l[i-1]+l[i]
        i+=1
    j=1
    while j<((n/2)+1):
        l[j]=s[j]
        l[-j-1]=s[j]
        j+=1
        
    n+=1
   
    print(l)
    print('....')
  
print(l)



```

```python
#fibonacci
l=[1,1]
n=1
while n<15:
    l.append(l[n-1]+l[n])
    n+=1
print(l)    




```
