```python
num = input("Inserire il numero di elementi per la serie di Fibonacci: ")
try:
    n = int(num)
    n0 = 1
    print(n0, end = " ")
    n1 = 1
    print(n1, end = " ")
    for k in range(2,n):
        n2 = n0 + n1
        print(n2, end = " ")
        n0 = n1
        n1 = n2
except:
    print("La prossima volta inserisci un numero")
```
