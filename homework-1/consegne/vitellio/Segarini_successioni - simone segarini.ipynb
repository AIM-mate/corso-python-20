# Successione di Fibonacci
Stampa i primi n numeri della successione di Fibonacci.

```python
n = 20

a0 = 0
a1 = 1

fib = [a0, a1]

if n == 1:
    print(fib[0])
elif n == 2:
    print(fib)

elif n > 2:
    for i in range(2, n):
        temp = a1 + a0
        a0 = a1
        a1 = temp
        fib.append(a1)
    print(fib)
else:
    print('n deve essere maggiore di zero!')
```

# Triangolo di Tartaglia
Scrivere le prime n righe del triangolo di Tartaglia 

```python
n = 10

a0 = [1]
a1 = [1, 1]

if n == 1:
    print(a0)
elif n == 2:
    print(a0)
    print(a1)
elif n > 2:
    print(a0)
    print(a1)
    for i in range(2, n):
        temp = []
        temp.append(1)
        for j in range(1, i):
            temp.append(a1[j-1] + a1[j])
        temp.append(1)
        print(temp)
        a1 = temp
```
