```python
inp=input("how many fibonacci's serie numbers do you want?\n")
n=int(inp)
v = [1 for x in range(n)]


for w in range(1,n-1,1):
    v[w+1]=v[w]+v[w-1]

    
print (v)
```

```python
inp=input("how many serie numbers do you want?\n")
n=int(inp)
v = [0 for x in range(n)]

v[0]=1
v[1]=7
v[2]=18

for w in range(2,n-1):
    v[w+1]=v[w]+v[w-2]-v[w-1]+2

    
print (v)
```

```python
inp=input("how many lines of Tartaglia triangle do you want?\n")
m=int(inp)
vold=[1]


for potenza in range(1,m+1):
    vnew = [1 for x in range(potenza)]
    vnew[0]=1
    vnew[potenza-1]=1
    for k in range(1,potenza-1):
        vnew[k]=vold[k-1]+vold[k]
    
    print(vnew)
    vold=vnew     
```
