```python
#Trasponi la lista di liste delle azioni (invertendo righe e colonne)
print("Stock_Value: ", Stocks_Value)
print('\n')
transpose=[[row[i] for row in Stocks_Value] for i in range(len(Stocks_Value[1]))]
print("Trasposta: ", transpose)
```

```python
#Trova i valori attesi (media) dei prezzi di ognuna delle 4 azioni 
mean=[sum(row)/len(row) for row in transpose]
print("Vettore delle medie: ", mean)
```

```python
#Trova le deviazioni standard dei prezzi di ognuna delle 4 azioni
dev_st = [[sum((transpose[j][i]-mean[j])**2 for i in range(len(transpose[j])))**(1/2)]for j in range(len(mean))]
print("Vettore delle deviazioni standard: ", dev_st)
```

```python
#Salva in una lista di liste tutte le possibili correlazioni tra i prezzi di ognuna delle 4 azioni 
#rho_xy= cov(x,y)/(dev_st_x*dev_stand_y)
corr = [[sum([(transpose[j][i]-mean[j])*(transpose[k][i]-mean[k]) for i in range(len(transpose[j]))])/(dev_st[j][0]*dev_st[k][0]) for k in range(len(mean))]for j in range(len(mean))]
print(corr)
```
