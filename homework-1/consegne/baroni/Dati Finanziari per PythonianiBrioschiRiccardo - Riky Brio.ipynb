**FACOLTATIVO** Implementa un CRR (albero binomiale) in grado di prezzare un'opzione call con scadenza 01 / 11 / 2020 per ognuna di queste 4 azioni, stimando ove possibile i parametri necessari. - _tante righe di codice_

```python
##Trasponi la lista di liste delle azioni (invertendo righe e colonne) - 1 riga di codice
ST=[[val[i] for val in Stocks_Value] for i in range(4)]
print(ST)
```

```python
##Trova i valori attesi (media) dei prezzi di ognuna delle 4 azioni - 1 riga di codice
Medie=[sum(val for val in ST[i])/len(ST[i])for i in range(len(ST))]
```

```python
##Trova le deviazioni standard dei prezzi di ognuna delle 4 azioni - 1 riga di codice
Dev=[[(sum((numero-Medie[i])**2 for numero in ST[i])/len(ST[i]))**(0.5)]for i in range(len(ST))]      
print(Dev)
```

```python
##Salva in una lista di liste tutte le possibili correlazioni tra i prezzi di ognuna delle 4 azioni - 1 riga di codice
```

```python
##Trova e salva per ogni giorno il nome dell'azienda la cui azione porterebbe ad un profitto maggiore o una perdita minore (in termini percentuali) se acquistata quel giorno e rivenduta il 2 Ottobre - 1 riga di codice
andamento=[[(Stocks_Value[len(Stocks_Value)-1][i]-Stocks_Value[j][i])/Stocks_Value[j][i] for i in range(4)]for j in range(len(Dates))]
indice=[andamento[i].index(max(andamento[i])) for i in range(len(andamento))]
Bestazienda=[Stocks_Name[i] for i in indice]
print(Bestazienda)
```

```python
##Generalizza il punto 5 anche per giorni precedenti al 2 Ottobre - 1 riga di codice
data=input("Inserisci data:\n")
index=Dates.index(data)
andamento=[[(Stocks_Value[index-1][i]-Stocks_Value[j][i])/Stocks_Value[j][i] for i in range(4)]for j in range(index)]
indice=[andamento[i].index(max(andamento[i])) for i in range(len(andamento))]
Bestazienda=[Stocks_Name[i] for i in indice]
print(Bestazienda)
```

```python
##Ripeti il punto 5 nel caso in cui ti interessi identificare il profitto minore o la perdita maggiore - 1 riga di codice
andamento=[[(Stocks_Value[len(Stocks_Value)-1][i]-Stocks_Value[j][i])/Stocks_Value[j][i] for i in range(4)]for j in range(len(Dates))]
indice=[andamento[i].index(min(andamento[i])) for i in range(len(andamento))]
WorstAzienda=[Stocks_Name[i] for i in indice]
print(WorstAzienda)
```
