```python
# esercizio1_Baroni
print(Stocks_Value)
print('\n')
transpose=[[row[i] for row in Stocks_Value] for i in range(len(Stocks_Value[1]))]
print(transpose)
```

```python
# esercizio2_Baroni

mean=[sum(row)/len(row) for row in transpose]

print(mean)
```

```python
# esercizio3_Baroni

dev_st=[sum((transpose[j][i]-mean[j])**2 for i in range(len(transpose[j])))**(1/2) for j in range(len(mean))]

print(dev_st)
```

```python
# esercizio4_Baroni 

# cov(x,y)/(dvst(x)*dvst(y))

corr=[[sum([(transpose[j][i]-mean[j])*(transpose[k][i]-mean[k]) for i in range(len(transpose[j]))])/(dev_st[j]*dev_st[k])for k in range(len(mean))]for j in range(len(mean))]
print(corr)

```
