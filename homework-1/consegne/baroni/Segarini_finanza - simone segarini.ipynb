```python
#1
Stocks_values_trasp = [[Stocks_Value[j][i] for j in range(len(Stocks_Value))] for i in range(len(Stocks_Value[0]))]
print(Stocks_values_trasp)
```

```python
#2
Estimated_values = []

for i in range(len(Stocks_values_trasp)):
    somma = 0
    for j in range(len(Stocks_values_trasp[0])):
        somma += Stocks_values_trasp[i][j]
    Estimated_values.append((somma/len(Stocks_values_trasp[0])))
print(Estimated_values)
```

```python
#3
sd = []

n = len(Stocks_values_trasp)
m = len(Stocks_values_trasp[0])

for i in range(n):
    temp = 0
    for j in range(m):
        temp += (Stocks_values_trasp[i][j] - Estimated_values[i])**2
    sd.append((temp/(m-1))**(1/2))
print(sd)
```

```python
#4
cor = []

n = len(Stocks_values_trasp)
m = len(Stocks_values_trasp[0])

for i in range(n):
    cor_temp = []
    for j in range(n):
        cov = 0
        for h in range(m):
            cov += (Stocks_values_trasp[i][h] - Estimated_values[i])*(Stocks_values_trasp[j][h] - Estimated_values[j])
        cor_temp.append(cov/((m-1)*sd[i]*sd[j]))
    cor.append(cor_temp)
print(cor)
```

```python
#5
aziende_top = []

n = len(Stocks_Value)
m = len(Stocks_Value[0])

for i in range(n):
    azienda_migliore = None
    guadagno_azione = []
    for j in range(m):
        guadagno_azione.append((Stocks_Value[-1][j] - Stocks_Value[i][j])/(Stocks_Value[i][j]))
    guadagno_max = max(guadagno_azione)
    for j in range(m):
        if(guadagno_azione[j] == guadagno_max):
            azienda_migliore = Stocks_Name[j]
            break
    aziende_top.append(azienda_migliore)
print(aziende_top)
```

```python
#6
aziende_top_dbd = []

n = len(Stocks_Value)
m = len(Stocks_Value[0])

for k in range(n):
    aziende_top = []
    i = 0
    while i < k:
        azienda_migliore = None
        guadagno_azione = []
        for j in range(m):
            guadagno_azione.append((Stocks_Value[k][j] - Stocks_Value[i][j])/(Stocks_Value[i][j]))
        guadagno_max = max(guadagno_azione)
        for j in range(m):
            if(guadagno_azione[j] == guadagno_max):
                azienda_migliore = Stocks_Name[j]
                break
        aziende_top.append(azienda_migliore)
        i += 1
    aziende_top_dbd.append(aziende_top)
print(aziende_top_dbd)
```
