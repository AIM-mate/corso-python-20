# Trova la Moneta Falsa

In un numpy array di n elementi (2 < n < 240) ci sono tutte monete di un certo peso ed una leggermente più pesante. Dovete scoprire quale è l'indice della moneta più pesante!
Per farlo dovete costruire una funzione TEST che dati due array confronti le loro somme e restituisca:
- 1 se il primo è più pesante
- 2 se il secondo è più pesante
- 0 se hanno lo stesso peso

Questa funzione è l'unico modo in cui potete accedere agli elementi dell'array ed il vostro obiettivo è scoprire la moneta falsa in meno TEST possibili.

```python
import numpy as np

def TEST(x,y):
    if sum(x)/len(x)>sum(y)/len(y):
        return 1
    elif sum(x)/len(x)<sum(y)/len(y):
        return 2
    return 0

def binary_search(arr, start, end, counter):
    if start == end:
        return [start, counter]
    arr_split = np.split(arr,[int((end-start)/2)+1])
    test = TEST(arr_split[0], arr_split[1])
    counter += 1
    if test == 1:
        return binary_search(arr_split[0], start, start + len(arr_split[0]) - 1, counter)
    if test == 2:
        return binary_search(arr_split[1], end - len(arr_split[1]) + 1, end, counter)
```

Il programma dovrà generare nel main una serie di numpy array di varie dimensione e ogni elemento inizializzato ad un numero casuale tra 10000 e 20000 e successivamente scgeliere un indice casuale e aggiungere 1 a quell'elemento.
Create poi una matrice contenente:
- Numero di elementi dell'i-esimo array
- Indice della moneta più pesante dell'i-esimo array
- Indice restituito dal vostro algoritmo dell'i-esimo array
- Numero di volte in cui avete usato TEST per l'i-esimo array

EXTRA BONUS: fatelo nel caso in cui non sapete se la moneta falsa è più pesante o leggera

```python
#BONUS

import numpy as np

def TEST_bonus(x,y):
    xval = float(sum(x))/float(len(x))
    yval = float(sum(y))/float(len(y))
    if xval.is_integer() == 0:
        return 1
    elif yval.is_integer() == 0:
        return 2
    if len(x) == 1 and len(y) == 1:
        return 0
    if len(x) == 1:
        return 1
    if len(y) == 1:
        return 2

def binary_search_bonus(arr, start, end, counter):
    if start == end:
        return [start, counter, 0]
    arr_split = np.split(arr,[int((end-start)/2)+1])
    test = TEST_bonus(arr_split[0], arr_split[1])
    counter += 1
    if test == 1:
        return binary_search_bonus(arr_split[0], start, start + len(arr_split[0]) - 1, counter)
    if test == 2:
        return binary_search_bonus(arr_split[1], end - len(arr_split[1]) + 1, end, counter)
    if test == 0:
        return [start, counter, 1]
```

```python
n = 4
dimensions = np.random.randint(2, 241, size = n)
values = np.random.randint(10000, 20001, size = n)
sol_M = np.zeros((n,4))
purses = []
indices = []

for i in range(n):
    purses.append(np.full((dimensions[i],1), values[i]))
    indices.append(np.random.randint(dimensions[i]))
    if np.random.randint(2):
        purses[i][indices[i]] += 1
    else:
        purses[i][indices[i]] -= 1
    sol_M[i][0] = dimensions[i]
    sol_M[i][1] = indices[i]

    sol = binary_search_bonus(purses[i],0,dimensions[i]-1,0)
    if sol[2]:
        arr_s = np.split(purses[i], [sol[0]+1])
        test = TEST_bonus(arr_s[0], arr_s[1])
        sol[1] += 1
        if test == 2:
            sol[0] += 1
    sol_M[i][2] = sol[0]
    sol_M[i][3] = sol[1]

print(sol_M)
```
