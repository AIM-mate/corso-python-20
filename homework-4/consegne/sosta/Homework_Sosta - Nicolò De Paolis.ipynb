```python
def TEST(v,w):
    if(np.sum(v) > np.sum(w)):
        return 1
    elif(np.sum(v) < np.sum(w)):
        return 2
    else:
        return 0
```

```python
import numpy as np
k = input('Qaunti vettori vuoi creare?? \n')
M = np.ndarray((int(k),4))
for a in range(int(k)):
    n = np.random.randint(2,240)
    x = np.random.randint(10000,20000)
    v = [x]*n
    i = np.random.randint(n)
    v[i] = v[i] + 1
    new_v = v 
    l = len(new_v)
    j = 0
    c = 0
    while(l>1):
        c = c + 1
        if(l%2 == 1):
            T = TEST(new_v[:l//2],new_v[l//2:l-1])
            if(T == 0):
                false_coin_index = j + l-1
                break
            elif(T == 1):
                new_v = new_v[:l//2]
            else:
                new_v = new_v[l//2:l-1]
                j = l//2 + j
        else:
            T = TEST(new_v[:l//2],new_v[l//2:])
            if(T == 1):
                new_v = new_v[:l//2]
            else:
                new_v = new_v[l//2:] 
                j = l//2 + j
        l = len(new_v)  
        false_coin_index = j
    print('La moneta falsa è la numero: ',false_coin_index+1)
    M[a] = [n, i, false_coin_index, c]
print (M)
```
