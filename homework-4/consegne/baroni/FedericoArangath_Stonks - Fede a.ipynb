3. [1pt] Converti il dataframe in numpy, calcola i ritorni giornalieri (prezzo_giorno_successivo/prezzo_giorno_precendete - 1) e visualizzali in 16 plot diversi 

```python
import numpy as np

dfnp = df.to_numpy()
rows, columns=dfnp.shape
day_returns=np.zeros((rows, columns))

for j in range(columns):
    for i in range(rows-1):
        day_returns[i,j]=dfnp[i+1,j]/dfnp[i, j]-1
    plt.plot(day_returns[:,j], label=f'{df.columns[j]} day returns')
    plt.legend()
    plt.show()
```

```python
day_returns_mean=np.mean(day_returns, axis=0)
day_returns_std=np.std(day_returns, axis=0)
day_returns_corr=np.corrcoef(day_returns)

print(df.columns[(np.where(day_returns_mean==np.max(day_returns_mean)))[0][0]])
print(df.columns[(np.where(day_returns_std==np.max(day_returns_std)))[0][0]])


```
