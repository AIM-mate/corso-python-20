3. [1pt] Converti il dataframe in numpy, calcola i ritorni giornalieri (prezzo_giorno_successivo/prezzo_giorno_precendete - 1) e visualizzali in 16 plot diversi 

```python
import numpy as np
import matplotlib.pyplot as plt

dfnp = df.to_numpy()
dfnp=dfnp.transpose()
dfnp=np.delete(dfnp,0,0)
dfnp
ret=np.zeros((16,1266))
for j in range(16):
    for i in range(1266):
        ret[j][i]=dfnp[j][i+1]/dfnp[j][i]-1
    plt.plot(ret[j])
    plt.show()
print(ret)

```
