2. [1pt] Scrivi una funzione che consenta di visualizzare il grafico dell'andamento del prezzo di un'azione, possibilmente in un intervallo di tempo compreso tra due date

```python
import matplotlib.pyplot as plt

def plotStock(df, stock, date1, date2):
    if date1 in list(df.loc[:, "Date"]):
        position = [df.loc[i, "Date"] == date1 for i in df.index]
        start = [i for i, x in enumerate(position) if x][0]
    else:
        start = 0
    if date2 in list(df.loc[:, "Date"]):
        position = [df.loc[i, "Date"] == date2 for i in df.index]
        end = [i for i, x in enumerate(position) if x][-1]
    else:
        end = len(df.index) - 1
    if end <= start or stock not in df.columns or stock == "Date":
        return False
    ax = plt.subplots()[1]
    ax.plot([df.loc[i, stock] for i in range(start, end + 1)])
    return True
```

3. [1pt] Converti il dataframe in numpy, calcola i ritorni giornalieri (prezzo_giorno_successivo/prezzo_giorno_precendete - 1) e visualizzali in 16 plot diversi 

```python
import numpy as np

dfnp = df.to_numpy()
df_returns = []
for i in range(1,len(df.columns)):
    df_returns.append(np.array([dfnp[j+1][i] - dfnp[j][i] for j in range(len(df.index) - 1)]))
    ax = plt.subplots()[1]
    plt.title(df.columns[i])
    ax.plot(df_returns[i-1], label = df.columns[i])
    ax.plot(np.zeros(len(df.index) - 1))
```
