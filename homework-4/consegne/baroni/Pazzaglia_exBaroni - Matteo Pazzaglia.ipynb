2. [1pt] Scrivi una funzione che consenta di visualizzare il grafico dell'andamento del prezzo di un'azione, possibilmente in un intervallo di tempo compreso tra due date

```python
import matplotlib.pyplot as plt
import matplotlib.dates as dates

def plotStock(df,datainizio,datafine,azione):
    df=df[(pd.to_datetime(df.Date) >= datainizio) & (pd.to_datetime(df.Date) <= datafine)]
    fig = plt.figure(dpi=128, figsize=(10,6))
    new_x = dates.datestr2num(df.Date)
    plt.plot(dates.num2date(new_x),df[azione])
    plt.title(azione)
    plt.ylabel('price')
    plt.xlabel('date')
    
    
    
    
```

```python
# BONUS -> frontiera ottima di portafoglio
# torniamo al df e lo "pulisco"

dati=df.iloc[1039:,1:]
dati.head()
dati.iloc[0,:]
```

```python
dati.head()
```

```python
# correlation matrix
corr_matrix = dati.pct_change().apply(lambda x: np.log(1+x)).corr()

#Covariance matrix
cov_matrix = dati.pct_change().apply(lambda x: np.log(1+x)).cov()
```

```python
p_ret = [] # portfolio returns
p_vol = [] # portfolio volatility
p_weights = [] # weights

assets = len(dati.columns)
iteration = 10000

# log yearly returns

logret=dfnp[1266,1:] / dfnp[1039,1:]
logret=np.log(logret.astype('float64'))
print(logret)

ann_sd = dati.pct_change().apply(lambda x: np.log(1+x)).std().apply(lambda x: x*np.sqrt(227)) #227 sono i giorni che ho fino oggi
```

```python
logret
```

```python
for pf in range(iteration):
    weights = np.random.random(assets) # Parto da pesi random
    weights = weights/np.sum(weights) # to sum 1
    p_weights.append(weights)
    ann_randomret = np.dot(weights, logret) # prodotto dei pesi e del rendimento nell'ultimo anno
    p_ret.append(ann_randomret) # aggiungo i rendimenti al mio array
    var = cov_matrix.mul(weights, axis=0).mul(weights, axis=1).sum().sum()# Portfolio Variance
    sd = np.sqrt(var) # standard deviation annuale
    ann_sd = sd*np.sqrt(227) # Annual standard deviation = volatilità
    p_vol.append(ann_sd) # aggiungo deviazione standard annuale

    
    
portafogli=pd.DataFrame({'Returns':p_ret, 'Volatility':p_vol})
portafogli.plot.scatter(x='Volatility', y='Returns', marker='o', s=3, alpha=0.3, grid=True, figsize=[10,10])
```

```python
# Posso considerare il portafoglio ottimo come quello che ha il miglior sharpe ratio

ptf = {'Returns':p_ret, 'Volatility':p_vol} #dizionario

for counter, symbol in enumerate(dati.columns.tolist()):
    ptf[symbol+' weight'] = [w[counter] for w in p_weights]  # salvo i pesi
portfolios  = pd.DataFrame(ptf) #DataFrame
 

rf = 0.01 # risk factor
optimal_port = portafogli.iloc[((portafogli['Returns']-rf)/portafogli['Volatility']).idxmax()]
optimal_port # cerco in portfolios la riga con questo returns e questa volatilità

portfolios[(portfolios['Returns'] == optimal_port[0]) & (portfolios['Volatility'] == optimal_port[1])]
```
