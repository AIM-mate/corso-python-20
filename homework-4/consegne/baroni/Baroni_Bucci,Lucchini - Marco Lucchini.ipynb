2. [1pt] Scrivi una funzione che consenta di visualizzare il grafico dell'andamento del prezzo di un'azione, possibilmente in un intervallo di tempo compreso tra due date

```python
df=df.set_index('Date')
```

```python
import matplotlib.pyplot as plt

# def plotStock( ...

def plotStocks(n_azione, data_i, data_f):
    
    azione=df[n_azione]
    plt.plot(azione[data_i:data_f],label = n_azione)
    plt.legend()
    plt.show

plotStocks('Amazon.com Inc (AMZN)','2017-01-07','2019-05-06')
```
