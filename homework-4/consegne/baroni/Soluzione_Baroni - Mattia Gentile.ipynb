2. [1pt] Scrivi una funzione che consenta di visualizzare il grafico dell'andamento del prezzo di un'azione, possibilmente in un intervallo di tempo compreso tra due date

```python
import matplotlib.pyplot as plt

def plotStock(name,date1,date2):
    dff=df.iloc[df[df['Date']==date1].index.astype(int)[0]:df[df['Date']==date2].index.astype(int)[0]]
    dff[name].plot()
    plt.title(name)

d1=('1-Jan-20')
d2=('13-Nov-20')
df=df.astype({'Date': 'string'})
plotStock('TESLA Inc.',d1,d2)
```

3. [1pt] Converti il dataframe in numpy, calcola i ritorni giornalieri (prezzo_giorno_successivo/prezzo_giorno_precendete - 1) e visualizzali in 16 plot diversi 

```python
import numpy as np
import matplotlib.pyplot as plt

dfnp = df.to_numpy()
x=range(len(dfnp)-1)
fig, axs=plt.subplots(4, 4, sharey=True, figsize=(20,20))
y=np.zeros((16,len(dfnp)-1))
for i in range(16):
    for j in range(len(dfnp)-1):
        y[i,j]=dfnp[j+1][i+1]/dfnp[j][i+1]-1
    axs[i//4][i%4].plot(x,y[i])
    axs[i//4][i%4].set_title(df.columns[i+1])
```

GOOGL 3 azioni\

Il valore del portafoglio è 3 * prezzo di GOOGL + 4 * prezzo di AMZN + 5 * prezzo di FB

Costruisci il portafoglio che fa stonks nel 2020, ovvero che, se costruito il 01/01/2020 con 1000000€, ad oggi avrebbe il maggior valore, con le seguenti condizioni:

- all'inizio in ogni società non puoi investire più di 200000€ (20%)
- il numero di azioni acquistate di ogni società deve essere un intero non negativo
- non ci sono tassi di interesse, dividendi, commissioni e simili
- bisogna investire almeno il 99% della liquidità, la rimenente non verrà considerat

```python
# Esempio di portafoglio:

portfolio_test = pd.DataFrame(data={'Code': ["GOOGL","AMZN","FB"], 'NumStocks': [3,4,5],
                                    'Value 01/01/20': [1193.6639,1646.7944,182.9187],
                                    'Value 13/11/20': [1501.4839,2650.7724,234.636]})

print(portfolio_test)
```

```python
start=df[df['Date']=='1-Jan-20'].to_numpy()
end=df[df['Date']=='13-Nov-20'].to_numpy()
unitary_costs=np.zeros(16)
costs=start[0,1:17]
for i in range(16):
    unitary_costs[i]=(end[0,i+1]-costs[i])/costs[i]

total_cost=0
j=0
indexes=[]
num=[]
while total_cost<800000:
    i=np.argmax(unitary_costs)
    indexes.append(i)
    num.append(200000//costs[i])
    unitary_costs[i]=0
    total_cost+=num[j]*costs[i]
    j+=1
while total_cost<1000000:
    i=np.argmax(unitary_costs)
    if unitary_costs[i]<=0:
        break
    if costs[i]<1000000-total_cost:
        indexes.append(i)
        num.append((1000000-total_cost)//costs[i])
        unitary_costs[i]=0
        total_cost+=num[j]*costs[i]
        j+=1
    else:
        unitary_costs[i]=0

for k in range(len(indexes)):
    print(df.columns[indexes[k]+1])
    print(num[k])
    k+=1
    
print("The total investment is: ")
print(total_cost)

starting_values=[]
ending_values=[]
for h in range(len(indexes)):
    starting_values.append(start[0,indexes[h]+1])
    ending_values.append(end[0,indexes[h]+1])

portfolio=pd.DataFrame(data={'Code': ["TESLA","AMZN","MSFT","FB","GOOGL","VLS.MI"], 'NumStocks': num[0:k+1],
                                    'Value 01/01/20': starting_values, 'Value 13/11/20': ending_values})
print(portfolio)
```
