3. [1pt] Converti il dataframe in numpy, calcola i ritorni giornalieri (prezzo_giorno_successivo/prezzo_giorno_precendete - 1) e visualizzali in 16 plot diversi 

```python
import numpy as np
import matplotlib.pyplot as plt

dfnp = df.to_numpy()
fine = len(dfnp)
j=1
rit = np.zeros(fine)
while j<16:
    rit = np.zeros(fine)
    i=0
    while i<fine-1:
        rit[i] = dfnp[i+1,j]/dfnp[i,j] - 1
        i = i+1
    print("Grafico numero",j)
    plt.plot(rit)
    j=j+1
    plt.show()
```
