3. [1pt] Converti il dataframe in numpy, calcola i ritorni giornalieri (prezzo_giorno_successivo/prezzo_giorno_precendete - 1) e visualizzali in 16 plot diversi 

```python
import numpy as np
from datetime import datetime as dt

dfnp = df.iloc[:,1:].to_numpy(dtype = np.float64) 
r = dfnp[1:,:]-dfnp[:-1,:]-1

dates_raw = df.iloc[:,0].to_numpy()
dates = [dt.strptime(dates_raw[i],'%d-%b-%y') for i in range(len(dates_raw))]

fig = plt.figure()
fig.set_figheight(60)
fig.set_figwidth(60)
for i in range(r.shape[1]):
    plt.subplot(4,4,i+1)
    plotStock(r[:,i],'4-Jan-16', '13-Nov-20')

plt.show()
```

4. [1pt] Rispetto ai ritorni, calcola la media, deviazione standard, le correlazioni. 

Quale azione ha ritorni medi maggiori da inizio Gennaio 2016 ad ora? Qual è la più volatile (maggiore deviazione standard)?\
Quale azione ha ritorni medi maggiori da inizio Gennaio 2020 ad ora? Qual è la più volatile (maggiore deviazione standard)?

```python
mean_r = np.mean(r, axis = 0)
std_r = np.std(r, axis = 0)
corr_r = np.corrcoef(r, rowvar = True)

max_mean16 = np.where(mean_r == max(mean_r))[0][0]
print('Azione con ritorni maggiori da Gennaio 2016 ad ora: ', df.columns[max_mean16+1])

max_std16 = np.where(std_r == max(std_r))[0][0]
print('\nAzione più volatile da Gennaio 2016 ad ora: ', df.columns[max_std16+1])

inGen20 = np.where(dates_raw == '1-Jan-20')[0][0]
mean_r_20 = np.mean(r[inGen20:,:], axis = 0)
std_r_20 = np.std(r[inGen20:,:], axis = 0)

max_mean20 = np.where(mean_r_20 == max(mean_r_20))[0][0]
print('\nAzione con ritorni maggiori da Gennaio 2020 ad ora: ', df.columns[max_mean20+1])

max_std20 = np.where(std_r_20 == max(std_r_20))[0][0]
print('\nAzione più volatile da Gennaio 2020 ad ora: ', df.columns[max_std20+1])
```
