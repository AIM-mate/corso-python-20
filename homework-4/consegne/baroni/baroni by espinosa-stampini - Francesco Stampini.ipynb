2. [1pt] Scrivi una funzione che consenta di visualizzare il grafico dell'andamento del prezzo di un'azione, possibilmente in un intervallo di tempo compreso tra due date

```python
import datetime
import matplotlib.pyplot as plt
df['Date'] = pd.to_datetime(df['Date'])

def plotStock(df, company, start_date, end_date):
    mask = (df['Date'] > start_date) & (df['Date'] <= end_date)
    newp = df.loc[mask]
    plt.plot(newp['Date'], newp[company],color='orange')
    plt.xticks(rotation='vertical',color='black')
    plt.yticks(color='black')
    plt.title(company+' shares from ' + str(start_date) + ' to ' + str(end_date), color='orange')
    
# provo a "teslarlo" ;-P
start=datetime.datetime(2018,1,1)
end=datetime.datetime(2018,11,13)
plotStock(df, 'TESLA Inc.', start, end)
```

3. [1pt] Converti il dataframe in numpy, calcola i ritorni giornalieri (prezzo_giorno_successivo/prezzo_giorno_precendete - 1) e visualizzali in 16 plot diversi 

```python
import numpy as np

dfnp = df.to_numpy()
dim = dfnp.shape
returns = dfnp[1:dim[0],1:dim[1]]/dfnp[:dim[0]-1,1:dim[1]]-1

fig,axs=plt.subplots(4,4, figsize=(20, 20))
for i in range(0,16):
    axs[i%4,int(i/4)].plot(returns[:,i])
    axs[i%4,int(i/4)].set_title(df.columns[i+1])
```

4. [1pt] Rispetto ai ritorni, calcola la media, deviazione standard, le correlazioni. 

Quale azione ha ritorni medi maggiori da inizio Gennaio 2016 ad ora? Qual è la più volatile (maggiore deviazione standard)?\
Quale azione ha ritorni medi maggiori da inizio Gennaio 2020 ad ora? Qual è la più volatile (maggiore deviazione standard)?

```python
media=returns.mean(0)
stdev=np.std(returns, axis=0, dtype=np.float64)

print('L\'azione che ha ritorni medi maggiori da inizio Gen 2016 e: '+df.columns[np.argmax(media)+1])
print('L\'azione più volatile da inizio Gen 2016 e: '+df.columns[np.argmax(stdev)+1])

ind1gen=int(df[df['Date'] == datetime.datetime(2020,1,1)].index.values)
short_returns=returns[ind1gen:,:]
minimedia=short_returns.mean(0)
ministdev=np.std(short_returns,axis=0,dtype=np.float64)

print('L\'azione che ha ritorni medi maggiori da inizio Gen 2020 e: '+df.columns[np.argmax(minimedia)+1])
print('L\'azione più volatile da inizio Gen 2020 e: '+df.columns[np.argmax(ministdev)+1])
```
