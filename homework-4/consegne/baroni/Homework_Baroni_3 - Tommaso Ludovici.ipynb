2. [1pt] Scrivi una funzione che consenta di visualizzare il grafico dell'andamento del prezzo di un'azione, possibilmente in un intervallo di tempo compreso tra due date

```python
import matplotlib.pyplot as plt
import numpy as np

def plotStock(df, stock, start, end):
    start_index = df[df['Date'] == start].index
    end_index = df[df['Date'] == end].index
    sub_df = df[start_index[0] : end_index[0]+1]
    
    fig, axis = plt.subplots(1, 1, figsize = (32,8))
    axis.plot(sub_df['Date'].tolist(), sub_df[stock].tolist(), '-', linewidth = 3)
    
    plt.title(stock, fontsize = 30)
    plt.xticks(np.arange(0, len(sub_df['Date'].tolist()), 30), fontsize = 20, rotation = 90)
    plt.yticks(fontsize = 20)
    plt.show()
    
plotStock(df, 'Amazon.com Inc (AMZN)', '5-Jan-16', '13-Nov-20')
```

3. [1pt] Converti il dataframe in numpy, calcola i ritorni giornalieri (prezzo_giorno_successivo/prezzo_giorno_precendete - 1) e visualizzali in 16 plot diversi 

```python
dates = df['Date'].tolist()
stocks = df.columns.tolist()[1:]
dfnp = df.drop('Date', axis = 1).to_numpy()

daily_returns = np.zeros((dfnp.shape[0]-1, dfnp.shape[1]))
dfnp_prev = dfnp[:-1]
dfnp_foll = dfnp[1:]

daily_returns = dfnp_foll/dfnp_prev -1

fig, axis = plt.subplots(len(stocks), 1, sharex = True, figsize = (40,128))
for i in range(len(stocks)):
    axis[i].plot(dates[1:], daily_returns.T[i], '-', linewidth = 2)
    axis[i].set_title(stocks[i], fontsize = 30)
    axis[i].set_xticks(np.arange(0, len(dates[1:]), 30))
    axis[i].set_xticklabels([dates[i] for i in np.arange(0, len(dates[1:]), 30)], rotation = 90, fontsize = 20)

plt.show()
print('test')
```

4. [1pt] Rispetto ai ritorni, calcola la media, deviazione standard, le correlazioni. 

Quale azione ha ritorni medi maggiori da inizio Gennaio 2016 ad ora? Qual è la più volatile (maggiore deviazione standard)?\
Quale azione ha ritorni medi maggiori da inizio Gennaio 2020 ad ora? Qual è la più volatile (maggiore deviazione standard)?

```python
def agg_since_date(dreturns, date):
    index = dates.index(date)
    media = np.mean(dreturns[index:], axis = 0)
    std_dev = np.std(dreturns[index:], axis = 0)
    corr = np.corrcoef(dreturns[index:])
    
    return [media, std_dev, corr]

since2016 = agg_since_date(daily_returns, '4-Jan-16')
best_stock = stocks[list(since2016[0]).index(since2016[0].max())]
vol_stock = stocks[list(since2016[1]).index(since2016[1].max())]
print(f"L'azione con i ritorni medi maggiori dal 2016 è:\n{best_stock}\nL'azione più volatile dal 2016 è:\n{vol_stock}\n")

since2020 = agg_since_date(daily_returns, '1-Jan-20')
best_stock = stocks[list(since2020[0]).index(since2020[0].max())]
vol_stock = stocks[list(since2020[1]).index(since2020[1].max())]
print(f"L'azione con i ritorni medi maggiori dal 2020 è:\n{best_stock}\nL'azione più volatile dal 2020 è:\n{vol_stock}\n")
```
