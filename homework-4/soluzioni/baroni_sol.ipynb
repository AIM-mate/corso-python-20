5. [1pt] Un portafoglio è un insieme di assets, ad esempio:

GOOGL 3 azioni\
AMZN 4 azioni\
FB 5 azioni

Il valore del portafoglio è 3 * prezzo di GOOGL + 4 * prezzo di AMZN + 5 * prezzo di FB

Costruisci il portafoglio che fa stonks nel 2020, ovvero che, se costruito il 01/01/2020 con 1000000€, ad oggi avrebbe il maggior valore, con le seguenti condizioni:

- all'inizio in ogni società non puoi investire più di 200000€ (20%)
- il numero di azioni acquistate di ogni società deve essere un intero non negativo
- non ci sono tassi di interesse, dividendi, commissioni e simili
- bisogna investire almeno il 99% della liquidità, la rimenente non verrà considerata

```python
startLiquidity = 1e6
liquidity = startLiquidity
limitPerStock = 0.2*startLiquidity
mininvestment = 0.99*startLiquidity

cols = ['Name','NumStocks','Value 01/01/20','Value 13/11/20']

portfolio = pd.DataFrame(columns = cols)

valuesStart = df.loc["1-Jan-20"].to_numpy()
valuesEnd = df.loc[df.index[-1]].to_numpy()

netReturns = valuesEnd / valuesStart -1

for bestReturn in np.sort(netReturns)[-5:]:
    bestIdx = np.where(netReturns == bestReturn)
    toadd = pd.DataFrame(
            [[df.columns[bestIdx][0],
              np.floor(limitPerStock/valuesStart[bestIdx])[0],
              valuesStart[bestIdx][0],
              valuesEnd[bestIdx][0]]],
            columns=cols)
    portfolio = portfolio.append(toadd)

print(portfolio)
```

```python
# Is it already over the min?

sum(portfolio["NumStocks"].to_numpy()*portfolio["Value 01/01/20"].to_numpy()) > mininvestment
```

```python
# removes the excess, alternatively uses smaller and smallers stocks to fill the remaining (doable with while loop)

portfolio["NumStocks"].iloc[0] = portfolio["NumStocks"].iloc[0] - np.floor((sum(portfolio["NumStocks"].to_numpy()*portfolio["Value 01/01/20"].to_numpy()) - mininvestment)/(portfolio["Value 01/01/20"].to_numpy()[0]))
```

```python
print(portfolio)
print("At start €",sum(portfolio["NumStocks"].to_numpy()*portfolio["Value 01/01/20"].to_numpy()))
print("At end €",sum(portfolio["NumStocks"].to_numpy()*portfolio["Value 13/11/20"].to_numpy()))
```
